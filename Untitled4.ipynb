```python
# @title
import os
import sys
import subprocess
import time
import numpy as np
import traci
import traci.constants as tc
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
import math
import matplotlib.pyplot as plt
# ==============================================================================
# 0. COLAB SETUP (Run on Colab platform)
# ==============================================================================
# The setup commands are moved to the __main__ block for automatic execution.
# ------------------------------------------------------------------------------


# ==============================================================================
# 1. HYPERPARAMETERS AND CONSTANTS
# ==============================================================================
# --- A2C Hyperparameters (TUNED FOR STABILITY) ---
STATE_DIM = 32
PHASE_DIM = 4  # Discrete number of green phases
HIDDEN_1 = 128
HIDDEN_2 = 64
ACTION_CONTINUOUS_DIM = 1
# IMPORTANT: These define the strict output range [10, 80]
MIN_DURATION = 10.0 # seconds (Strict Minimum)
MAX_DURATION = 80.0 # seconds (Strict Maximum)
# **TUNED** Lower LR for stability against divergence
LR = 35e-6
GAMMA = 0.99
# **TUNED** Increased ENTROPY_BETA to force more exploration (break gridlock)
ENTROPY_BETA = 0.003

# --- SUMO Configuration ---
SUMO_CFG = "_1.sumocfg"
SIM_DURATION_STEPS = 1500 # 1 hour of simulation time steps per episode
MAX_EPISODES = 300 # Keep this for now, but you plan to run to 750+
# **TUNED** Increased batch size for better gradient estimation on GPU
TRAINING_BATCH_SIZE = 128

# --- Environment Parameters (ADJUST THESE FOR YOUR NETWORK!) ---
TLS_ID = "C"

# Corrected incoming lanes and detectors based on previous working cells
INCOMING_LANES = [
    "N2C_0", "N2C_1", # North Approach
    "E2C_0", "E2C_1", # East Approach
    "S2C_0", "S2C_1", # South Approach
    "W2C_0", "W2C_1"  # West Approach
]

INCOMING_DETECTORS = [
    "flow-N2C-0", "flow-N2C-1", # North Detectors
    "flow-E2C-0", "flow-E2C-1", # East Detectors
    "flow-S2C-0", "flow-S2C-1", # South Detectors
    "flow-W2C-0", "flow-W2C-1"  # West Detectors
]

# Phase Definitions (Must match the length and logic of your TLS)
PHASE_DEFINITIONS = [
    {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},  # Phase 0: N/S Straight
    {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},  # Phase 1: E/W Straight
    {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},  # Phase 2: N/S Left Turns
    {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}   # Phase 3: E/W Left Turns
]
ALL_RED_STRING = "r" * len(PHASE_DEFINITIONS[0]["green"])

# --- NEW REWARD CONFIGURATION ---
# R_wait: Penalty for absolute current waiting time (User requested to keep)
REWARD_WEIGHT_WAIT_TIME = 0.7
# R_duration: Bonus for longer green duration (towards 80s)
REWARD_WEIGHT_DURATION = 15.0
# R_diff: Reward for Queue Reduction (L_t - L_{t+1}) (NEW)
REWARD_WEIGHT_QUEUE_DIFF = 1.0
# R_bumper: Bonus for total queue < BUMPER_THRESHOLD (NEW)
BUMPER_REWARD_BONUS = 20.0
BUMPER_THRESHOLD = 4

global_sumo_process = None


# ==============================================================================
# 2. ACTOR-CRITIC NETWORK (From a2c_agent.py)
# ==============================================================================

class ActorCriticNet(nn.Module):
    """Hybrid Actor-Critic Network sharing a backbone for feature extraction."""
    def __init__(self, state_dim, phase_dim):
        super(ActorCriticNet, self).__init__()

        # --- Shared Feature Extractor Backbone ---
        self.backbone = nn.Sequential(
            nn.Linear(state_dim, HIDDEN_1),
            nn.ReLU(),
            nn.Linear(HIDDEN_1, HIDDEN_2),
            nn.ReLU()
        )

        # --- Actor Head ---
        self.actor_phase = nn.Linear(HIDDEN_2, phase_dim)
        self.actor_duration_mu = nn.Linear(HIDDEN_2, ACTION_CONTINUOUS_DIM)
        self.actor_duration_log_std = nn.Linear(HIDDEN_2, ACTION_CONTINUOUS_DIM)

        # --- Critic Head ---
        self.critic_value = nn.Linear(HIDDEN_2, 1)

    def forward(self, state):
        x = self.backbone(state)

        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = self.actor_duration_log_std(x)

        log_std = torch.clamp(log_std, min=-20, max=2)
        value = self.critic_value(x)

        return phase_logits, mu, log_std, value

def calculate_returns(rewards, values, dones, gamma):
    """Calculates discounted cumulative returns (TD Target) for a trajectory batch."""
    returns = []
    R = values[-1]

    for r, v, d in zip(reversed(rewards), reversed(values), reversed(dones)):
        if d:
            R = 0
        R = r + gamma * R
        returns.insert(0, R)

    return torch.tensor(returns).float()


# ==============================================================================
# 3. A2C AGENT CLASS (From a2c_agent.py)
# ==============================================================================

class A2CAgent:
    """The A2C agent orchestrates action selection and network learning."""
    def __init__(self, state_dim, phase_dim, lr=LR, gamma=GAMMA, checkpoint_path="a2c_checkpoint.pth"):

        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.gamma = gamma
        self.checkpoint_path = checkpoint_path
        self.training_batch_size = TRAINING_BATCH_SIZE

        self.net = ActorCriticNet(state_dim, phase_dim).to(self.device)
        self.optimizer = optim.Adam(self.net.parameters(), lr=lr)

        self.min_duration = MIN_DURATION
        self.max_duration = MAX_DURATION

        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': [], 'dones': [], 'actions': []}

    @torch.no_grad()
    def select_action(self, state):
        """Samples the hybrid action (phase and duration) from the policy."""
        self.net.eval()
        state_tensor = torch.tensor(state, dtype=torch.float32).to(self.device).unsqueeze(0)

        # 1. Forward Pass
        phase_logits, mu, log_std, value = self.net(state_tensor)

        # 2. Discrete Action (Phase) Sampling
        phase_dist = distributions.Categorical(logits=phase_logits)
        phase_action = phase_dist.sample()
        log_prob_phase = phase_dist.log_prob(phase_action)

        # 3. Continuous Action (Duration) Sampling
        std = torch.exp(log_std)
        duration_dist = distributions.Normal(mu, std)
        duration_action_raw = duration_dist.sample()

        # Clip the sampled duration to safety limits (guarantees output is in [10, 80])
        duration_action_clipped = torch.clamp(
            duration_action_raw,
            min=self.min_duration,
            max=self.max_duration
        )

        log_prob_duration = duration_dist.log_prob(duration_action_raw)

        # 4. Final Output
        log_prob_total = log_prob_phase + log_prob_duration.squeeze()

        # The actual chosen duration (integer for SUMO)
        chosen_duration = int(duration_action_clipped.item())

        final_action = (
            phase_action.item(),
            chosen_duration
        )

        self.net.train()
        return final_action, log_prob_total, value.squeeze(), chosen_duration


    def store_transition(self, state, reward, done, log_prob, value, chosen_duration):
        """Stores a single step of experience into the buffer."""
        self.buffer['states'].append(torch.tensor(state, dtype=torch.float32).to(self.device))
        self.buffer['rewards'].append(reward)
        self.buffer['values'].append(value.item())
        self.buffer['log_probs'].append(log_prob)
        self.buffer['dones'].append(done)
        self.buffer['actions'].append(chosen_duration)

    def compute_loss_and_update(self):
        """Calculates A2C losses for the entire buffer and updates the network weights."""
        if not self.buffer['states']:
            return 0.0

        # 1. Prepare Batch Tensors
        states = torch.stack(self.buffer['states']).to(self.device)
        rewards = self.buffer['rewards']
        values = torch.tensor(self.buffer['values']).float().to(self.device)
        log_probs = torch.stack(self.buffer['log_probs']).to(self.device)
        dones = self.buffer['dones']

        # 2. Calculate Returns (TD Target)
        returns = calculate_returns(rewards, values, dones, self.gamma).to(self.device)

        # 3. Calculate Advantage
        Advantage = returns - values

        # 4. CRITIC LOSS
        critic_loss = Advantage.pow(2).mean()

        # 5. ACTOR LOSS
        actor_loss = -(log_probs * Advantage.detach()).mean()

        # 6. ENTROPY LOSS
        phase_logits, mu, log_std, _ = self.net(states)
        phase_dist = distributions.Categorical(logits=phase_logits)
        entropy = phase_dist.entropy().mean()

        # Total Loss
        total_loss = critic_loss + actor_loss - ENTROPY_BETA * entropy

        # 7. Optimization Step
        self.optimizer.zero_grad()
        total_loss.backward()
        nn.utils.clip_grad_norm_(self.net.parameters(), max_norm=0.5)
        self.optimizer.step()

        # 8. Clear the buffer after update
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': [], 'dones': [], 'actions': []}

        return total_loss.item()

    def save_model(self):
        """Saves the network's state dictionary and optimizer state."""
        torch.save({
            'model_state_dict': self.net.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict(),
        }, self.checkpoint_path)
        print(f"Model saved to {self.checkpoint_path}")

    def load_model(self):
        """Loads the network's state dictionary and optimizer state."""
        try:
            checkpoint = torch.load(self.checkpoint_path, map_location=self.device)
            self.net.load_state_dict(checkpoint['model_state_dict'])
            self.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            print(f"Model loaded successfully from {self.checkpoint_path}")
        except FileNotFoundError:
            print("Checkpoint file not found. Starting training from scratch.")
        except Exception as e:
            print(f"Error loading model: {e}. Starting training from scratch.")

# ==============================================================================
# 4. TRAFFIC AGENT (SUMO Interface, from traffic_agent.py)
# ==============================================================================

class TrafficAgent:
    """Class to manage interaction between the RL Agent and the SUMO simulation via TraCI."""

    def __init__(self, tls_id, incoming_lanes, incoming_detectors, phase_definitions, all_red_string,
                 min_green_time=10,
                 yellow_time=2,
                 all_red_time=1,
                 reward_weight_wait_time=REWARD_WEIGHT_WAIT_TIME,
                 reward_weight_duration=REWARD_WEIGHT_DURATION,
                 reward_weight_queue_diff=REWARD_WEIGHT_QUEUE_DIFF,
                 bumper_reward_bonus=BUMPER_REWARD_BONUS,
                 bumper_threshold=BUMPER_THRESHOLD
                ):

        self.tls_id = tls_id
        self.incoming_lanes = incoming_lanes
        self.incoming_detectors = incoming_detectors
        self.phase_definitions = phase_definitions
        self.num_green_phases = len(phase_definitions)
        self.all_red_string = all_red_string

        self.min_green_time = min_green_time
        self.max_green_time = MAX_DURATION
        self.yellow_time = yellow_time
        self.all_red_time = all_red_time

        self.current_phase_index = 0
        self.current_phase_state = "GREEN"
        self.time_in_current_phase = 0
        self.next_phase_index = 0

        # Metrics for state vector (current observation L_{t+1})
        self.current_queue_lengths = np.zeros(len(self.incoming_detectors), dtype=float)
        self.current_waiting_times = np.zeros(len(self.incoming_lanes), dtype=float)

        # REWARD WEIGHTS
        self.reward_weight_wait_time = reward_weight_wait_time
        self.reward_weight_duration = reward_weight_duration
        self.reward_weight_queue_diff = reward_weight_queue_diff
        self.bumper_reward_bonus = bumper_reward_bonus
        self.bumper_threshold = bumper_threshold

        # Tracking for queue difference reward (Previous observation L_t)
        self.last_queue_lengths = np.zeros(len(self.incoming_detectors), dtype=float)


        # Initializing other inputs
        num_approaches = len(self.incoming_lanes) // 2
        self.priority_of_lanes = np.full(num_approaches, 1/num_approaches, dtype=float)
        self.avg_speeds = np.zeros(len(self.incoming_lanes), dtype=float)
        self.phase_one_hot = np.zeros(self.num_green_phases, dtype=float)

        # --- Tracking for Avg Green Time ---
        self.phase_durations = np.zeros(self.num_green_phases, dtype=int)
        self.phase_counts = np.zeros(self.num_green_phases, dtype=int)
        # ----------------------------------------


        print(f"Agent for {self.tls_id} initialized.")

        try:
            traci.trafficlight.setRedYellowGreenState(
                self.tls_id,
                self.phase_definitions[self.current_phase_index]["green"]
            )
        except traci.TraCIException as e:
            print(f"Error setting initial state: {e}")

    def get_state(self):
        """Collects the current state vector (L_{t+1})."""
        queue_lengths = []
        lane_speeds = []

        for det_id in self.incoming_detectors:
            try:
                queue = traci.lanearea.getLastStepHaltingNumber(det_id)
                queue_lengths.append(queue)
                current_lane_speed = traci.lanearea.getLastStepMeanSpeed(det_id)
                lane_speeds.append(current_lane_speed)
            except traci.TraCIException:
                queue_lengths.append(0)
                lane_speeds.append(0.0)

        waiting_times = []
        for lane_id in self.incoming_lanes:
            try:
                wait = traci.lane.getWaitingTime(lane_id)
                waiting_times.append(wait)
            except traci.TraCIException:
                waiting_times.append(0)


        # Store the latest metrics for reward calculation (L_{t+1})
        self.current_queue_lengths = np.array(queue_lengths, dtype=float)
        self.current_waiting_times = np.array(waiting_times, dtype=float)
        self.avg_speeds = np.array(lane_speeds, dtype=float)

        # One-hot encoding for the current green phase.
        self.phase_one_hot = np.zeros(self.num_green_phases, dtype=float)
        if self.current_phase_state == "GREEN":
            self.phase_one_hot[self.current_phase_index] = 1.0


        # Construct the 32-element state vector
        state_vector = np.concatenate([
            self.current_queue_lengths,
            self.current_waiting_times,
            self.priority_of_lanes,
            self.avg_speeds,
            self.phase_one_hot
        ])

        if len(state_vector) != STATE_DIM:
             print(f"WARNING: State dimension mismatch! Expected {STATE_DIM}, Got {len(state_vector)}")

        return state_vector

    def get_reward(self, chosen_duration):
        """
        Calculates the weighted sum reward based on:
        1. Absolute current waiting time (Penalty, R_wait).
        2. Queue length reduction (Reward, R_diff).
        3. Bumper reward for very low queues (Bonus, R_bumper).
        4. Duration encouragement (Bonus, R_duration).
        """

        # R_wait: Absolute Waiting Time Penalty (from current step L_{t+1})
        current_total_wait = np.sum(self.current_waiting_times)
        r_wait_penalty = -self.reward_weight_wait_time * current_total_wait

        # --- Queue-Based Rewards (R_diff and R_bumper) ---

        # R_diff: Reward for Queue Reduction (Q_t - Q_{t+1})
        Q_t = np.sum(self.last_queue_lengths)
        Q_t_plus_1 = np.sum(self.current_queue_lengths)

        queue_reduction = Q_t - Q_t_plus_1
        r_queue_diff_reward = self.reward_weight_queue_diff * queue_reduction

        # R_bumper: Bonus if final queue is below threshold
        r_bumper_bonus = 0.0
        if Q_t_plus_1 < self.bumper_threshold:
            r_bumper_bonus = self.bumper_reward_bonus

        # R_duration: DURATION ENCOURAGEMENT TERM (Intrinsic Reward)
        duration_range = MAX_DURATION - MIN_DURATION # 70
        duration_normalized = (chosen_duration - MIN_DURATION) / duration_range
        r_duration_bonus = duration_normalized * self.reward_weight_duration

        # Final Weighted Sum
        final_reward = r_wait_penalty + r_queue_diff_reward + r_bumper_bonus + r_duration_bonus

        return final_reward


    def step(self, action):
        # 1. Unpack the agent's decision
        next_green_phase_index, green_duration = action

        current_green_phase_index = self.current_phase_index
        self.next_phase_index = next_green_phase_index

        try:
            # 2. Set light to Yellow
            yellow_string = self.phase_definitions[current_green_phase_index]["yellow"]
            traci.trafficlight.setRedYellowGreenState(self.tls_id, yellow_string)
            self.current_phase_state = "YELLOW"
            self.time_in_current_phase = 0

            # 3. Advance the simulation for the fixed yellow time
            for _ in range(self.yellow_time):
                traci.simulationStep()
                self.time_in_current_phase += 1

            # 4. Set light to All-Red
            traci.trafficlight.setRedYellowGreenState(self.tls_id, self.all_red_string)
            self.current_phase_state = "ALL_RED"
            self.time_in_current_phase = 0

            # 5. Advance the simulation for the fixed all-red time
            for _ in range(self.all_red_time):
                traci.simulationStep()
                self.time_in_current_phase += 1

            # 6. Get the Green phase string for the NEXT phase
            green_string = self.phase_definitions[next_green_phase_index]["green"]

            # 7. Set light to the NEW Green phase
            traci.trafficlight.setRedYellowGreenState(self.tls_id, green_string)
            self.current_phase_state = "GREEN"
            self.current_phase_index = next_green_phase_index
            self.time_in_current_phase = 0

            # --- Tracking Green Time ---
            self.phase_durations[next_green_phase_index] += green_duration
            self.phase_counts[next_green_phase_index] += 1
            # --------------------------------

            # 8. Advance the simulation for the chosen duration
            for _ in range(green_duration):
                traci.simulationStep()
                self.time_in_current_phase += 1

        except traci.TraCIException as e:
            print(f"TraCI Error during step execution: {e}")

        # 9. Before getting new state, save the current queue length (L_t) as 'last'
        self.last_queue_lengths = self.current_queue_lengths.copy()

        # 10. Get the resulting state (L_{t+1})
        new_state = self.get_state()

        # 11. Calculate reward using L_t (last_queue_lengths) and L_{t+1} (current_queue_lengths)
        reward = self.get_reward(green_duration)

        done = False
        info = {}

        return new_state, reward, done, info

    def get_episode_metrics(self):
        """
        Calculates and returns key performance indicators for the current step.
        """

        try:
            total_halting = np.sum([traci.lanearea.getLastStepHaltingNumber(det_id) for det_id in self.incoming_detectors])
        except traci.TraCIException:
             total_halting = 0

        total_waiting_time = 0
        try:
            for lane_id in self.incoming_lanes:
                # Summing cumulative waiting time of all vehicles on the approach lanes
                total_waiting_time += traci.lane.getWaitingTime(lane_id)
        except traci.TraCIException:
            total_waiting_time = 0

        try:
            # Average speed across all monitored lanes
            avg_speed = np.mean([traci.lanearea.getLastStepMeanSpeed(det_id) for det_id in self.incoming_detectors])
        except traci.TraCIException:
            avg_speed = 0.0

        try:
            # Throughput: Total number of vehicles that *finished* their route in this step.
            throughput = traci.simulation.getDepartedNumber()
        except traci.TraCIException:
            throughput = 0

        # We also want the *total* number of vehicles in the network for a better picture
        try:
            total_vehicles = traci.vehicle.getIDCount()
        except traci.TraCIException:
            total_vehicles = 0

        return {
            'total_halting': total_halting,
            'total_wait_sec': total_waiting_time,
            'avg_speed': avg_speed,
            'throughput_step': throughput,
            'total_vehicles': total_vehicles
        }

    def get_avg_green_times(self):
        """Calculates the average green duration used for each phase in the episode."""
        avg_times = np.zeros(self.num_green_phases)
        for i in range(self.num_green_phases):
            if self.phase_counts[i] > 0:
                avg_times[i] = self.phase_durations[i] / self.phase_counts[i]
            else:
                avg_times[i] = 0
        return avg_times.round(2)


# ==============================================================================
# 5. SUMO CONTROL FUNCTIONS (From train_traffic_a2c.py)
# ==============================================================================

def stop_sumo():
    """Stops the SUMO simulation and closes the TraCI connection."""
    global global_sumo_process

    try:
        if traci.is_connected():
             traci.close()
    except:
        pass

    if global_sumo_process is not None:
        try:
            global_sumo_process.terminate()
            global_sumo_process.wait(timeout=5)
            print("Previous SUMO subprocess terminated.")
        except Exception:
            pass
        global_sumo_process = None


def start_sumo(sumo_binary, sumo_config_file):
    """Starts SUMO as a headless subprocess and connects TraCI (Colab optimized)."""
    global global_sumo_process

    # --- Robust Cleanup (NEW) ---
    print("Attempting to clean up previous TraCI connection and SUMO process...")
    stop_sumo()
    time.sleep(1) # Give the OS a moment to free the port
    # ----------------------------

    if 'SUMO_HOME' not in os.environ:
        print("ERROR: SUMO_HOME not set. Did you run the setup code?")
        return False

    SUMO_BINARY_PATH = os.path.join(os.environ['SUMO_HOME'], 'bin', 'sumo')

    sumo_cmd = [
        SUMO_BINARY_PATH,
        "-c", sumo_config_file,
        "--step-length", "1",
        "--remote-port", "8813",
        "--no-step-log",
        "--start"
    ]

    print(f"Starting SUMO (Headless) with command: {SUMO_BINARY_PATH} -c {sumo_config_file} ...")

    # Start the subprocess silently
    global_sumo_process = subprocess.Popen(sumo_cmd, shell=False, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(1)

    # TraCI connect attempt loop
    max_retries = 10
    for i in range(max_retries):
        try:
            traci.init(8813)
            print("TraCI connection established.")
            return True
        except traci.exceptions.TraCIException as e:
            if "Connection refused" in str(e):
                time.sleep(1) # Keep waiting if refused
            elif "Connection 'default' is already active" in str(e):
                 # This should now be handled by the stop_sumo() cleanup,
                 # but if it happens, we assume the previous process is dead
                 # and we wait to retry connecting.
                print("Warning: Detected active connection error, retrying...")
                stop_sumo() # Try cleaning up one more time just in case
                time.sleep(1)
            else:
                print(f"Error connecting to TraCI: {e}")
                return False
    print("Failed to connect to TraCI after multiple retries.")
    return False

# ==============================================================================
# 6. MAIN TRAINING LOGIC (MODIFIED FOR SPEED)
# ==============================================================================

def train_agent():
    """Main function to run the A2C training loop."""

    agent = A2CAgent(STATE_DIM, PHASE_DIM)
    episode_rewards = []

    # --- Training Progress Metrics ---
    episode_rewards_list = []
    avg_queue_lengths_list = []
    avg_waiting_times_list = []
    avg_speeds_list = []
    total_throughput_list = []
    avg_green_times_list = []

    # Start SUMO ONCE before the loop
    if not start_sumo(None, SUMO_CFG):
        return

    try:
        for episode in range(MAX_EPISODES):
            print(f"\n--- EPISODE {episode + 1}/{MAX_EPISODES} ---")

            # Reset simulation state via TraCI
            if episode > 0:
                traci.load(["-c", SUMO_CFG]) # Reload the configuration file
                traci.simulationStep()        # Advance one step to initialize traffic

            traffic_env = TrafficAgent(
                tls_id=TLS_ID,
                incoming_lanes=INCOMING_LANES,
                incoming_detectors=INCOMING_DETECTORS,
                phase_definitions=PHASE_DEFINITIONS,
                all_red_string=ALL_RED_STRING
            )

            state = traffic_env.get_state()
            total_reward = 0
            step_counter = 0

            # --- Episode-specific metric trackers ---
            episode_halting_sum = 0
            episode_wait_sum = 0
            episode_speed_sum = 0
            episode_throughput = 0
            num_steps_tracked = 0

            # Episode Simulation Loop
            while step_counter < SIM_DURATION_STEPS:

                # A. Agent selects action
                action, log_prob, value, chosen_duration = agent.select_action(state)

                # B. Environment executes action
                new_state, reward, done, _ = traffic_env.step(action)

                # C. Store experience
                agent.store_transition(state, reward, done, log_prob, value, chosen_duration)

                # D. Prepare for next step
                state = new_state
                total_reward += reward

                # --- Collect Step Metrics ---
                metrics = traffic_env.get_episode_metrics()
                episode_halting_sum += metrics['total_halting']
                episode_wait_sum += metrics['total_wait_sec']
                episode_speed_sum += metrics['avg_speed']
                episode_throughput += metrics['throughput_step']
                num_steps_tracked += 1

                # Advance step counter by the duration of the cycle executed
                step_counter += (traffic_env.yellow_time + traffic_env.all_red_time + action[1])

                # Update the network
                if len(agent.buffer['states']) >= agent.training_batch_size:
                    agent.compute_loss_and_update()

            # 5. End of Episode Update and Monitoring
            if agent.buffer['states']:
                agent.compute_loss_and_update()

            # --- MONITORING OUTPUT ---
            episode_rewards.append(total_reward)
            episode_rewards_list.append(total_reward)
            avg_reward = np.mean(episode_rewards[-20:])

            # Calculate average episode metrics
            if num_steps_tracked > 0:
                avg_queue_len = episode_halting_sum / num_steps_tracked
                avg_episode_speed = episode_speed_sum / num_steps_tracked
            else:
                avg_queue_len = 0.0
                avg_episode_speed = 0.0

            total_wait_time = episode_wait_sum
            avg_green_times = traffic_env.get_avg_green_times()

            avg_queue_lengths_list.append(avg_queue_len)
            avg_waiting_times_list.append(total_wait_time)
            avg_speeds_list.append(avg_episode_speed)
            total_throughput_list.append(episode_throughput)
            avg_green_times_list.append(avg_green_times)

            # Print episode summary
            print(f"EPISODE {episode + 1} FINISHED:")
            print(f"  Total Reward: {total_reward:.2f}")
            print(f"  Avg Reward (Last 20): {avg_reward:.2f}")
            print(f"  Avg Queue Length: {avg_queue_len:.2f} vehicles")
            print(f"  Total Wait Time: {total_wait_time:.2f} seconds")
            print(f"  Avg Speed: {avg_episode_speed:.2f} m/s")
            print(f"  Total Throughput: {episode_throughput} vehicles")
            print(f"  Avg Green Time (P0-P3): {avg_green_times}")
            print(f"  Simulation Time Advanced: {step_counter} seconds")

            # Save the model periodically
            if (episode + 1) % 10 == 0:
                agent.save_model()

    except traci.exceptions.TraCIException as e:
        print(f"Critical TraCI Error: {e}. Stopping training.")
    finally:
        stop_sumo()


    # --- PLOTTING RESULTS ---
    print("\n\n--- Generating Performance Plots ---")

    episodes = range(1, len(episode_rewards_list) + 1)

    # 5 AXES FOR BASE METRICS + 1 AXIS FOR AVG GREEN TIMES
    fig, axs = plt.subplots(6, 1, figsize=(10, 18), sharex=True)
    fig.suptitle('A2C Traffic Control Performance Tracking Over Episodes', fontsize=16)

    # 1. Total Reward
    axs[0].plot(episodes, episode_rewards_list, label='Total Episode Reward', color='green')
    axs[0].set_ylabel('Total Reward')
    axs[0].grid(True, linestyle='--')

    # 2. Average Queue Length
    axs[1].plot(episodes, avg_queue_lengths_list, label='Avg. Queue Length', color='red')
    axs[1].set_ylabel('Avg Queue Length (Vehicles)')
    axs[1].grid(True, linestyle='--')

    # 3. Total Waiting Time
    axs[2].plot(episodes, avg_waiting_times_list, label='Total Waiting Time', color='orange')
    axs[2].set_ylabel('Total Wait Time (sec)')
    axs[2].grid(True, linestyle='--')

    # 4. Average Speed
    axs[3].plot(episodes, avg_speeds_list, label='Avg. Speed', color='blue')
    axs[3].set_ylabel('Avg Speed (m/s)')
    axs[3].grid(True, linestyle='--')

    # 5. Throughput
    axs[4].plot(episodes, total_throughput_list, label='Total Throughput', color='purple')
    axs[4].set_ylabel('Total Throughput (Vehicles)')
    axs[4].grid(True, linestyle='--')

    # 6. Avg Green Time per Phase (NEW PLOT)
    avg_green_times_matrix = np.array(avg_green_times_list)

    if avg_green_times_matrix.shape[1] > 0:
        axs[5].plot(episodes, avg_green_times_matrix[:, 0], label='Phase 0 (N/S Straight)', linestyle='-', marker='.')
        axs[5].plot(episodes, avg_green_times_matrix[:, 1], label='Phase 1 (E/W Straight)', linestyle='-', marker='x')
        axs[5].plot(episodes, avg_green_times_matrix[:, 2], label='Phase 2 (N/S Left)', linestyle='--', marker='+')
        axs[5].plot(episodes, avg_green_times_matrix[:, 3], label='Phase 3 (E/W Left)', linestyle='--', marker='s')

        axs[5].set_ylabel('Avg Green Time (seconds)')
        axs[5].set_xlabel('Episode')
        axs[5].legend(loc='upper left', fontsize='small')
        axs[5].grid(True, linestyle='--')


    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    return agent


if __name__ == "__main__":

    # -------------------------------------------------------------
    # Colab-specific setup to ensure SUMO is installed and configured
    print("--- Running Colab Setup ---")

    # 2. Set environment variables
    os.environ['SUMO_HOME'] = "/usr/share/sumo"
    sys.path.append(os.path.join(os.environ['SUMO_HOME'], 'tools'))
    print(f"SUMO_HOME set to: {os.environ.get('SUMO_HOME')}")
    print("--- Setup Complete. Starting Training. ---")
    # -------------------------------------------------------------

    train_agent()
```

```python
# @title 5. A2C TRAINING with BEST MODEL SAVING & RELOADING
import os
import sys
import subprocess
import time
import numpy as np
import traci
import traci.constants as tc
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
import matplotlib.pyplot as plt

# ==============================================================================
# 0. SAVING & LOADING CONFIGURATION
# ==============================================================================
# Set this to True if you want to load the previous "best_model.pth" and continue
LOAD_EXISTING_MODEL = False
MODEL_DIR = "models"
BEST_MODEL_PATH = os.path.join(MODEL_DIR, "best_model.pth")
LATEST_MODEL_PATH = os.path.join(MODEL_DIR, "latest_model.pth")

# Create directory if it doesn't exist
os.makedirs(MODEL_DIR, exist_ok=True)

# ==============================================================================
# 1. HYPERPARAMETERS
# ==============================================================================
STATE_DIM = 32
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64
ACTION_CONTINUOUS_DIM = 1

# Bounds
MIN_DURATION = 20.0
MAX_DURATION = 80.0

# Learning Rates
# WARNING: 99e-4 is approx 0.01. This is very high.
# If training is unstable, revert to 3e-5.
LR = 99e-4
GAMMA = 0.99
ENTROPY_BETA = 0.003

# Reward Weights
REWARD_WEIGHT_WAIT_TIME = 0.7
REWARD_WEIGHT_DURATION = 15.0
REWARD_WEIGHT_QUEUE_DIFF = 1.0
BUMPER_REWARD_BONUS = 20.0
BUMPER_THRESHOLD = 4

# Simulation Config
SUMO_CFG = "_1.sumocfg"
SIM_DURATION_STEPS = 1500
MAX_EPISODES = 200
BATCH_SIZE = 128

TLS_ID = "C"
INCOMING_LANES = ["N2C_0", "N2C_1", "E2C_0", "E2C_1", "S2C_0", "S2C_1", "W2C_0", "W2C_1"]
INCOMING_DETECTORS = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]

PHASE_DEFINITIONS = [
    {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
    {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
    {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
    {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
]
ALL_RED_STRING = "r" * 12

global_sumo_process = None

# ==============================================================================
# 2. NEURAL NETWORK
# ==============================================================================
class ActorCriticNet(nn.Module):
    def __init__(self, state_dim, phase_dim):
        super(ActorCriticNet, self).__init__()
        self.backbone = nn.Sequential(
            nn.Linear(state_dim, HIDDEN_1),
            nn.ReLU(),
            nn.Linear(HIDDEN_1, HIDDEN_2),
            nn.ReLU()
        )
        self.actor_phase = nn.Linear(HIDDEN_2, phase_dim)
        self.actor_duration_mu = nn.Linear(HIDDEN_2, 1)
        self.actor_duration_log_std = nn.Linear(HIDDEN_2, 1)
        self.critic_value = nn.Linear(HIDDEN_2, 1)
        with torch.no_grad():
            self.actor_duration_mu.bias.data.fill_(35.0)
            self.actor_duration_log_std.bias.data.fill_(-1.0)

    def forward(self, state):
        x = self.backbone(state)
        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = self.actor_duration_log_std(x)
        log_std = torch.clamp(log_std, min=-20, max=2)
        value = self.critic_value(x)
        return phase_logits, mu, log_std, value

# ==============================================================================
# 3. AGENT CLASS (Updated with Save/Load)
# ==============================================================================
class A2CAgent:
    def __init__(self):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.net = ActorCriticNet(STATE_DIM, PHASE_DIM).to(self.device)
        self.optimizer = optim.Adam(self.net.parameters(), lr=LR)
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': [], 'dones': []}

    def select_action(self, state):
        self.net.eval()
        state_t = torch.tensor(state, dtype=torch.float32).to(self.device).unsqueeze(0)

        with torch.no_grad():
            phase_logits, mu, log_std, value = self.net(state_t)

        phase_dist = distributions.Categorical(logits=phase_logits)
        phase_action = phase_dist.sample()

        std = torch.exp(log_std)
        dur_dist = distributions.Normal(mu, std)
        dur_raw = dur_dist.sample()
        dur_clipped = torch.clamp(dur_raw, MIN_DURATION, MAX_DURATION)

        log_prob = phase_dist.log_prob(phase_action) + dur_dist.log_prob(dur_raw).squeeze()

        return (phase_action.item(), int(dur_clipped.item())), log_prob, value.squeeze()

    def update(self):
        if not self.buffer['states']: return

        states = torch.stack(self.buffer['states']).to(self.device)
        rewards = self.buffer['rewards']
        values = torch.tensor(self.buffer['values']).float().to(self.device)
        log_probs = torch.stack(self.buffer['log_probs']).to(self.device)

        returns = []
        R = values[-1]
        for r in reversed(rewards):
            R = r + GAMMA * R
            returns.insert(0, R)
        returns = torch.tensor(returns).float().to(self.device)

        adv = returns - values
        loss = (adv.pow(2).mean()) - (log_probs * adv.detach()).mean()

        logits, _, _, _ = self.net(states)
        entropy = distributions.Categorical(logits=logits).entropy().mean()
        loss -= ENTROPY_BETA * entropy

        self.optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(self.net.parameters(), 0.5)
        self.optimizer.step()
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': [], 'dones': []}

    # --- NEW SAVING/LOADING METHODS ---
    def save_model(self, path):
        torch.save({
            'model_state_dict': self.net.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict()
        }, path)

    def load_model(self, path):
        if os.path.exists(path):
            checkpoint = torch.load(path, map_location=self.device)
            self.net.load_state_dict(checkpoint['model_state_dict'])
            self.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            print(f" Model loaded successfully from {path}")
            return True
        else:
            print(f" Model file not found at {path}. Starting from scratch.")
            return False

# ==============================================================================
# 4. ENVIRONMENT
# ==============================================================================
class TrafficAgent:
    def __init__(self, tls_id):
        self.tls_id = tls_id
        self.lanes = INCOMING_LANES
        self.detectors = INCOMING_DETECTORS
        self.phases = PHASE_DEFINITIONS
        self.phase_idx = 0
        self.last_queue_sum = 0
        self.phase_durations = np.zeros(PHASE_DIM)
        self.phase_counts = np.zeros(PHASE_DIM)
        traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[0]["green"])

    def get_state(self):
        queues = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        waits = np.array([traci.lane.getWaitingTime(l) for l in self.lanes])
        speeds = np.array([traci.lanearea.getLastStepMeanSpeed(d) for d in self.detectors])
        phase_oh = np.zeros(PHASE_DIM)
        phase_oh[self.phase_idx] = 1.0
        priority = np.full(4, 0.25)
        return np.concatenate([queues, waits, priority, speeds, phase_oh])

    def step(self, action):
        next_phase, duration = action
        self.phase_durations[next_phase] += duration
        self.phase_counts[next_phase] += 1

        if self.phase_idx != next_phase:
            traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[self.phase_idx]["yellow"])
            for _ in range(2): traci.simulationStep()
            traci.trafficlight.setRedYellowGreenState(self.tls_id, ALL_RED_STRING)
            for _ in range(1): traci.simulationStep()

        traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[next_phase]["green"])
        for _ in range(duration): traci.simulationStep()

        self.phase_idx = next_phase
        state = self.get_state()

        current_queues = np.sum([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        current_wait = np.sum([traci.lane.getWaitingTime(l) for l in self.lanes])

        r_wait = -REWARD_WEIGHT_WAIT_TIME * current_wait
        r_diff = REWARD_WEIGHT_QUEUE_DIFF * (self.last_queue_sum - current_queues)
        r_bump = BUMPER_REWARD_BONUS if current_queues < BUMPER_THRESHOLD else 0.0
        norm_dur = (duration - MIN_DURATION) / (MAX_DURATION - MIN_DURATION)
        r_dur = norm_dur * REWARD_WEIGHT_DURATION

        final_reward = r_wait + r_diff + r_bump + r_dur
        self.last_queue_sum = current_queues

        return state, final_reward

    def get_avg_green_times(self):
        return np.divide(self.phase_durations, self.phase_counts, out=np.zeros_like(self.phase_durations), where=self.phase_counts!=0).round(2)

# ==============================================================================
# 5. MAIN TRAINING LOOP (With Saving Logic)
# ==============================================================================
def train_agent():
    global global_sumo_process

    # History
    history = {'rewards': [], 'queues': [], 'waits': [], 'speeds': [], 'throughput': [], 'greens': []}

    # Best Model Tracking
    best_avg_reward = -float('inf')

    # Start SUMO
    if global_sumo_process: global_sumo_process.kill()
    sumo_cmd = ["sumo", "-c", SUMO_CFG, "--step-length", "1", "--no-step-log", "--remote-port", "8813"]
    global_sumo_process = subprocess.Popen(sumo_cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(2)
    traci.init(8813)

    agent = A2CAgent()

    # --- LOAD MODEL IF REQUESTED ---
    if LOAD_EXISTING_MODEL:
        if agent.load_model(BEST_MODEL_PATH):
            print(">>> Resuming training from BEST model.")
        else:
            print(">>> Could not load model. Training from scratch.")

    try:
        print(f"--- STARTING TRAINING (Saving to: {MODEL_DIR}/) ---")
        for episode in range(MAX_EPISODES):
            if episode > 0: traci.load(["-c", SUMO_CFG]); traci.simulationStep()

            env = TrafficAgent(TLS_ID)
            state = env.get_state()
            ep_metrics = {'rew': 0, 'halt': 0, 'wait': 0, 'speed': 0, 'thru': 0, 'steps': 0}
            step_counter = 0

            while step_counter < SIM_DURATION_STEPS:
                action, log_prob, value = agent.select_action(state)
                next_state, reward = env.step(action)

                agent.buffer['states'].append(torch.tensor(state, dtype=torch.float32))
                agent.buffer['rewards'].append(reward)
                agent.buffer['values'].append(value.item())
                agent.buffer['log_probs'].append(log_prob)

                state = next_state
                step_counter += (3 + action[1])

                ep_metrics['rew'] += reward
                ep_metrics['halt'] += np.sum([traci.lanearea.getLastStepHaltingNumber(d) for d in INCOMING_DETECTORS])
                ep_metrics['wait'] += np.sum([traci.lane.getWaitingTime(l) for l in INCOMING_LANES])
                ep_metrics['speed'] += np.mean([traci.lanearea.getLastStepMeanSpeed(d) for d in INCOMING_DETECTORS])
                ep_metrics['thru'] += traci.simulation.getDepartedNumber()
                ep_metrics['steps'] += 1

                if len(agent.buffer['states']) >= BATCH_SIZE:
                    agent.update()

            agent.update()

            # Stats
            avg_rew = ep_metrics['rew']
            avg_q = ep_metrics['halt'] / ep_metrics['steps']
            tot_wait = ep_metrics['wait']
            avg_spd = ep_metrics['speed'] / ep_metrics['steps']
            tot_thru = traci.vehicle.getIDCount()
            avg_greens = env.get_avg_green_times()

            history['rewards'].append(avg_rew)
            history['queues'].append(avg_q)
            history['waits'].append(tot_wait)
            history['speeds'].append(avg_spd)
            history['throughput'].append(tot_thru)
            history['greens'].append(avg_greens)

            # --- SAVING LOGIC ---
            # 1. Calculate Moving Average Reward (last 10 episodes) for stability
            recent_rewards = history['rewards'][-10:]
            current_avg_score = sum(recent_rewards) / len(recent_rewards)

            # 2. Check if this is the best model so far
            saved_msg = ""
            if current_avg_score > best_avg_reward and episode > 10:
                best_avg_reward = current_avg_score
                agent.save_model(BEST_MODEL_PATH)
                saved_msg = " [ NEW BEST SAVED]"

            # 3. Always save the latest model (just in case)
            agent.save_model(LATEST_MODEL_PATH)

            print(f"EPISODE {episode + 1}: Reward {avg_rew:.0f} | Avg(10): {current_avg_score:.0f}{saved_msg} | Wait {tot_wait:.0f}")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        traci.close()
        if global_sumo_process: global_sumo_process.kill()

    # Plots
    print("\n--- Generating Performance Plots ---")
    episodes = range(1, len(history['rewards']) + 1)
    fig, axs = plt.subplots(6, 1, figsize=(10, 20), sharex=True)
    axs[0].plot(episodes, history['rewards'], color='green'); axs[0].set_ylabel('Total Reward')
    axs[1].plot(episodes, history['queues'], color='red'); axs[1].set_ylabel('Avg Queue Length')
    axs[2].plot(episodes, history['waits'], color='orange'); axs[2].set_ylabel('Total Wait (sec)')
    axs[3].plot(episodes, history['speeds'], color='blue'); axs[3].set_ylabel('Avg Speed (m/s)')
    axs[4].plot(episodes, history['throughput'], color='purple'); axs[4].set_ylabel('Total Throughput')
    greens = np.array(history['greens'])
    if len(greens) > 0:
        markers = ['.', 'x', '+', 's']
        for i in range(4):
            axs[5].plot(episodes, greens[:, i], label=f'Phase {i}', marker=markers[i], markersize=4, linestyle='-')
        axs[5].legend()
        axs[5].set_ylabel('Avg Green (s)')
    axs[5].set_xlabel('Episode')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    train_agent()
```

```python
#v3 a2c
# @title 8. FULL TWO-STAGE SYSTEM (With Complete Graphing)
import os
import sys
import subprocess
import time
import numpy as np
import traci
import traci.constants as tc
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
import matplotlib.pyplot as plt
import copy
import shutil

# ==============================================================================
# CONFIGURATION
# ==============================================================================
# --- Phase 1: Scouting (High LR) ---
PHASE_1_LR = 99e-4  # User Requested High LR
PHASE_1_EPISODES = 150
TOP_K = 5

# --- Phase 2: Refinement (Stable LR) ---
PHASE_2_LR = 5e-5
PHASE_2_EPISODES = 100

# Common Config
STATE_DIM = 32
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64
GAMMA = 0.99
ENTROPY_BETA = 0.005
MIN_DURATION = 10.0
MAX_DURATION = 80.0
BATCH_SIZE = 256
SIM_DURATION_STEPS = 1500

# File Paths
BASE_DIR = "two_stage_experiment_graphs"
STAGE_1_DIR = os.path.join(BASE_DIR, "stage1_scouting")
STAGE_2_DIR = os.path.join(BASE_DIR, "stage2_refined")

os.makedirs(STAGE_1_DIR, exist_ok=True)
os.makedirs(STAGE_2_DIR, exist_ok=True)

# SUMO Config
SUMO_CFG = "_1.sumocfg"
TLS_ID = "C"
INCOMING_LANES = ["N2C_0", "N2C_1", "E2C_0", "E2C_1", "S2C_0", "S2C_1", "W2C_0", "W2C_1"]
INCOMING_DETECTORS = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]
PHASE_DEFINITIONS = [
    {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
    {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
    {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
    {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
]
ALL_RED_STRING = "r" * 12
global_sumo_process = None

# ==============================================================================
# UTILS: PLOTTING & SAVING
# ==============================================================================
def save_full_graphs(history, title, save_path):
    """Generates the 6-subplot figure for any given history dict."""
    episodes = range(1, len(history['rewards']) + 1)
    fig, axs = plt.subplots(6, 1, figsize=(10, 18), sharex=True)
    fig.suptitle(title, fontsize=16)

    # 1. Rewards
    axs[0].plot(episodes, history['rewards'], color='green')
    axs[0].set_ylabel('Total Reward')
    axs[0].grid(True, linestyle='--')

    # 2. Queues
    axs[1].plot(episodes, history['queues'], color='red')
    axs[1].set_ylabel('Avg Queue')
    axs[1].grid(True, linestyle='--')

    # 3. Wait Time
    axs[2].plot(episodes, history['waits'], color='orange')
    axs[2].set_ylabel('Total Wait (s)')
    axs[2].grid(True, linestyle='--')

    # 4. Speed
    axs[3].plot(episodes, history['speeds'], color='blue')
    axs[3].set_ylabel('Avg Speed (m/s)')
    axs[3].grid(True, linestyle='--')

    # 5. Throughput
    axs[4].plot(episodes, history['throughput'], color='purple')
    axs[4].set_ylabel('Throughput')
    axs[4].grid(True, linestyle='--')

    # 6. Green Times
    greens = np.array(history['greens'])
    if len(greens) > 0:
        markers = ['.', 'x', '+', 's']
        for i in range(4):
            axs[5].plot(episodes, greens[:, i], label=f'Phase {i}', marker=markers[i])
        axs[5].legend(loc='upper left')
        axs[5].set_ylabel('Avg Green (s)')
        axs[5].grid(True, linestyle='--')

    axs[5].set_xlabel('Episode')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    print(f"   >>> Saving Plot to: {save_path}")
    plt.savefig(save_path)
    plt.close()

class TopKSaver:
    def __init__(self, k, save_dir):
        self.k = k
        self.save_dir = save_dir
        self.top_models = []

    def update(self, score, episode, model):
        state = copy.deepcopy(model.state_dict())
        self.top_models.append((score, episode, state))
        self.top_models.sort(key=lambda x: x[0], reverse=True)
        self.top_models = self.top_models[:self.k]

    def save_to_disk(self):
        print(f"\n>>> Saving Top {self.k} Candidates to {self.save_dir}...")
        for rank, (score, ep, state) in enumerate(self.top_models):
            filename = f"rank_{rank+1}_score_{int(score)}_ep_{ep}.pth"
            path = os.path.join(self.save_dir, filename)
            torch.save(state, path)
        return self.top_models

# ==============================================================================
# AGENT & ENV
# ==============================================================================
class ActorCriticNet(nn.Module):
    def __init__(self):
        super(ActorCriticNet, self).__init__()
        self.backbone = nn.Sequential(nn.Linear(STATE_DIM, HIDDEN_1), nn.ReLU(), nn.Linear(HIDDEN_1, HIDDEN_2), nn.ReLU())
        self.actor_phase = nn.Linear(HIDDEN_2, PHASE_DIM)
        self.actor_duration_mu = nn.Linear(HIDDEN_2, 1)
        self.actor_duration_log_std = nn.Linear(HIDDEN_2, 1)
        self.critic_value = nn.Linear(HIDDEN_2, 1)
        with torch.no_grad():
            self.actor_duration_mu.bias.data.fill_(35.0)
            self.actor_duration_log_std.bias.data.fill_(-1.0)

    def forward(self, state):
        x = self.backbone(state)
        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = torch.clamp(self.actor_duration_log_std(x), -20, 2)
        value = self.critic_value(x)
        return phase_logits, mu, log_std, value

class A2CAgent:
    def __init__(self, lr):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.net = ActorCriticNet().to(self.device)
        self.optimizer = optim.Adam(self.net.parameters(), lr=lr)
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': []}

    def select_action(self, state):
        self.net.eval()
        t_state = torch.tensor(state, dtype=torch.float32).to(self.device).unsqueeze(0)
        with torch.no_grad():
            p_logits, mu, log_std, val = self.net(t_state)

        p_dist = distributions.Categorical(logits=p_logits)
        phase = p_dist.sample()
        d_dist = distributions.Normal(mu, torch.exp(log_std))
        dur_raw = d_dist.sample()
        dur = torch.clamp(dur_raw, MIN_DURATION, MAX_DURATION)

        log_prob = p_dist.log_prob(phase) + d_dist.log_prob(dur_raw).squeeze()
        return (phase.item(), int(dur.item())), log_prob, val.squeeze()

    def update(self):
        if not self.buffer['states']: return
        states = torch.stack(self.buffer['states']).to(self.device)
        rewards = self.buffer['rewards']
        values = torch.tensor(self.buffer['values']).float().to(self.device)
        log_probs = torch.stack(self.buffer['log_probs']).to(self.device)

        returns = []
        R = values[-1]
        for r in reversed(rewards):
            R = r + GAMMA * R
            returns.insert(0, R)
        returns = torch.tensor(returns).float().to(self.device)

        adv = returns - values
        loss = adv.pow(2).mean() - (log_probs * adv.detach()).mean()
        logits, _, _, _ = self.net(states)
        loss -= ENTROPY_BETA * distributions.Categorical(logits=logits).entropy().mean()

        self.optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(self.net.parameters(), 0.5)
        self.optimizer.step()
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': []}

class TrafficAgent:
    def __init__(self):
        self.last_q = 0
        traci.trafficlight.setRedYellowGreenState(TLS_ID, PHASE_DEFINITIONS[0]["green"])
        self.phase_idx = 0
        self.phase_durations = np.zeros(PHASE_DIM)
        self.phase_counts = np.zeros(PHASE_DIM)

    def get_state(self):
        qs = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in INCOMING_DETECTORS])
        ws = np.array([traci.lane.getWaitingTime(l) for l in INCOMING_LANES])
        sps = np.array([traci.lanearea.getLastStepMeanSpeed(d) for d in INCOMING_DETECTORS])
        ph = np.zeros(PHASE_DIM); ph[self.phase_idx] = 1.0
        return np.concatenate([qs, ws, np.full(4, 0.25), sps, ph])

    def step(self, action):
        next_p, dur = action
        self.phase_durations[next_p] += dur
        self.phase_counts[next_p] += 1

        if self.phase_idx != next_p:
            traci.trafficlight.setRedYellowGreenState(TLS_ID, PHASE_DEFINITIONS[self.phase_idx]["yellow"])
            for _ in range(2): traci.simulationStep()
            traci.trafficlight.setRedYellowGreenState(TLS_ID, ALL_RED_STRING)
            for _ in range(1): traci.simulationStep()

        traci.trafficlight.setRedYellowGreenState(TLS_ID, PHASE_DEFINITIONS[next_p]["green"])
        for _ in range(dur): traci.simulationStep()
        self.phase_idx = next_p

        q = np.sum([traci.lanearea.getLastStepHaltingNumber(d) for d in INCOMING_DETECTORS])
        w = np.sum([traci.lane.getWaitingTime(l) for l in INCOMING_LANES])

        rew = (-0.7 * w) + (1.0 * (self.last_q - q)) + (20.0 if q < 4 else 0) + (15.0 * ((dur-10)/70))
        self.last_q = q
        return self.get_state(), rew

    def get_metrics(self):
        qs = np.sum([traci.lanearea.getLastStepHaltingNumber(d) for d in INCOMING_DETECTORS])
        ws = np.sum([traci.lane.getWaitingTime(l) for l in INCOMING_LANES])
        sp = np.mean([traci.lanearea.getLastStepMeanSpeed(d) for d in INCOMING_DETECTORS])
        th = traci.vehicle.getIDCount()
        gr = np.divide(self.phase_durations, self.phase_counts, out=np.zeros_like(self.phase_durations), where=self.phase_counts!=0)
        return qs, ws, sp, th, gr

# ==============================================================================
# EXECUTION
# ==============================================================================
def start_sumo():
    global global_sumo_process
    if global_sumo_process: global_sumo_process.kill()
    sumo_cmd = ["sumo", "-c", SUMO_CFG, "--step-length", "1", "--no-step-log", "--remote-port", "8813"]
    global_sumo_process = subprocess.Popen(sumo_cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(2)
    traci.init(8813)

def run_episode_full_stats(agent, env):
    state = env.get_state()
    metrics = {'rew': 0, 'q': 0, 'w': 0, 's': 0, 't': 0, 'steps': 0}

    steps = 0
    while steps < SIM_DURATION_STEPS:
        action, log_prob, val = agent.select_action(state)
        next_state, reward = env.step(action)

        agent.buffer['states'].append(torch.tensor(state, dtype=torch.float32))
        agent.buffer['rewards'].append(reward)
        agent.buffer['values'].append(val.item())
        agent.buffer['log_probs'].append(log_prob)

        state = next_state
        steps += (3 + action[1])

        # Accumulate stats
        q, w, s, t, gr = env.get_metrics()
        metrics['rew'] += reward
        metrics['q'] += q
        metrics['w'] += w
        metrics['s'] += s
        metrics['t'] = t # Use current total count
        metrics['steps'] += 1

        if len(agent.buffer['states']) >= BATCH_SIZE: agent.update()
    agent.update()

    return {
        'rew': metrics['rew'],
        'q': metrics['q']/metrics['steps'],
        'w': metrics['w'],
        's': metrics['s']/metrics['steps'],
        't': metrics['t'],
        'gr': gr
    }

# --- PHASE 1 ---
def run_phase_1():
    print(f"\n{'='*40}\n PHASE 1: SCOUTING (LR={PHASE_1_LR})\n{'='*40}")
    agent = A2CAgent(lr=PHASE_1_LR)
    saver = TopKSaver(k=TOP_K, save_dir=STAGE_1_DIR)
    history = {'rewards': [], 'queues': [], 'waits': [], 'speeds': [], 'throughput': [], 'greens': []}

    start_sumo()
    try:
        for ep in range(PHASE_1_EPISODES):
            if ep > 0: traci.load(["-c", SUMO_CFG]); traci.simulationStep()
            env = TrafficAgent()
            stats = run_episode_full_stats(agent, env)

            history['rewards'].append(stats['rew'])
            history['queues'].append(stats['q'])
            history['waits'].append(stats['w'])
            history['speeds'].append(stats['s'])
            history['throughput'].append(stats['t'])
            history['greens'].append(stats['gr'])

            saver.update(stats['rew'], ep+1, agent.net)
            print(f"P1 | Ep {ep+1}: Rew {stats['rew']:.0f} | Q {stats['q']:.1f} | Wait {stats['w']:.0f}")

    finally:
        traci.close()
        saver.save_to_disk()
        save_full_graphs(history, f"Phase 1 Scouting Results (LR={PHASE_1_LR})", os.path.join(STAGE_1_DIR, "phase1_overview.png"))

    return saver.top_models

# --- PHASE 2 ---
def run_phase_2(top_models_data):
    print(f"\n{'='*40}\n PHASE 2: REFINEMENT (LR={PHASE_2_LR})\n{'='*40}")

    for rank, (score, source_ep, state_dict) in enumerate(top_models_data):
        print(f"\n>>> Refining Rank {rank+1} (Source Score: {score:.0f})")

        agent = A2CAgent(lr=PHASE_2_LR)
        agent.net.load_state_dict(state_dict)
        history = {'rewards': [], 'queues': [], 'waits': [], 'speeds': [], 'throughput': [], 'greens': []}

        start_sumo()
        try:
            for ep in range(PHASE_2_EPISODES):
                if ep > 0: traci.load(["-c", SUMO_CFG]); traci.simulationStep()
                env = TrafficAgent()
                stats = run_episode_full_stats(agent, env)

                history['rewards'].append(stats['rew'])
                history['queues'].append(stats['q'])
                history['waits'].append(stats['w'])
                history['speeds'].append(stats['s'])
                history['throughput'].append(stats['t'])
                history['greens'].append(stats['gr'])

                print(f"   Rank {rank+1} | Ep {ep+1}: Rew {stats['rew']:.0f} | Q {stats['q']:.1f}")

            # Save Model & Graphs for this specific candidate
            model_path = os.path.join(STAGE_2_DIR, f"Rank_{rank+1}_Refined.pth")
            graph_path = os.path.join(STAGE_2_DIR, f"performance_rank_{rank+1}.png")

            torch.save(agent.net.state_dict(), model_path)
            save_full_graphs(history, f"Refinement: Rank {rank+1} (Orig Score {score:.0f})", graph_path)

        finally:
            traci.close()

if __name__ == "__main__":
    top_candidates = run_phase_1()
    run_phase_2(top_candidates)
    print("\n DONE. Check 'two_stage_experiment_graphs' folder for plots.")
```

```python
#v4 a2c # @title 11. COMPLETE 3-TIER DIAGNOSTIC (Queue + Wait + Phases)
import os
import sys
import torch
import traci
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ==============================================================================
# CONFIG
# ==============================================================================
MODEL_PATH = "two_stage_experiment_graphs/stage2_refined/Rank_4_Refined.pth"
SUMO_BINARY = "sumo"
SUMO_CFG = "_1.sumocfg"

# Network Params
STATE_DIM = 32
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64

# ==============================================================================
# MODEL
# ==============================================================================
class ActorCriticNet(torch.nn.Module):
    def __init__(self):
        super(ActorCriticNet, self).__init__()
        self.backbone = torch.nn.Sequential(
            torch.nn.Linear(STATE_DIM, HIDDEN_1),
            torch.nn.ReLU(),
            torch.nn.Linear(HIDDEN_1, HIDDEN_2),
            torch.nn.ReLU()
        )
        self.actor_phase = torch.nn.Linear(HIDDEN_2, PHASE_DIM)
        self.actor_duration_mu = torch.nn.Linear(HIDDEN_2, 1)
        self.actor_duration_log_std = torch.nn.Linear(HIDDEN_2, 1)
        self.critic_value = torch.nn.Linear(HIDDEN_2, 1)

    def forward(self, state):
        x = self.backbone(state)
        return self.actor_phase(x), self.actor_duration_mu(x), self.actor_duration_log_std(x), self.critic_value(x)

# ==============================================================================
# VISUALIZATION
# ==============================================================================
def visualize_complete_picture():
    # 1. Load Model
    device = torch.device("cpu")
    model = ActorCriticNet().to(device)

    if not os.path.exists(MODEL_PATH):
        print(f" Error: Could not find model at {MODEL_PATH}")
        return

    model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
    model.eval()

    # 2. Start Simulation
    try: traci.close()
    except: pass
    traci.start([SUMO_BINARY, "-c", SUMO_CFG, "--step-length", "1", "--no-step-log"])

    # 3. Lanes Grouped by Direction (For Waiting Time)
    lanes_N = ["N2C_0", "N2C_1"]
    lanes_E = ["E2C_0", "E2C_1"]
    lanes_S = ["S2C_0", "S2C_1"]
    lanes_W = ["W2C_0", "W2C_1"]
    all_detectors = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]

    # 4. History Container
    history = {
        'step': [], 'phase': [],
        'q_N': [], 'q_E': [], 'q_S': [], 'q_W': [],
        'w_N': [], 'w_E': [], 'w_S': [], 'w_W': [] # Wait times
    }

    # 5. Run
    step = 0
    tls_id = "C"
    phases = [
        {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
        {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
        {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
        {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
    ]
    curr_phase = 0
    traci.trafficlight.setRedYellowGreenState(tls_id, phases[0]["green"])

    print(">>> Collecting Queue AND Wait Time Data...")

    while traci.simulation.getMinExpectedNumber() > 0 and step < 1500:
        # Helper to Record Data
        def record(ph):
            # Queues (Halting)
            qN = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in lanes_N])
            qE = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in lanes_E])
            qS = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in lanes_S])
            qW = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in lanes_W])

            # Wait Time (Accumulated Waiting Time)
            wN = np.sum([traci.lane.getWaitingTime(l) for l in lanes_N])
            wE = np.sum([traci.lane.getWaitingTime(l) for l in lanes_E])
            wS = np.sum([traci.lane.getWaitingTime(l) for l in lanes_S])
            wW = np.sum([traci.lane.getWaitingTime(l) for l in lanes_W])

            history['step'].append(step)
            history['phase'].append(ph)
            history['q_N'].append(qN); history['q_E'].append(qE); history['q_S'].append(qS); history['q_W'].append(qW)
            history['w_N'].append(wN); history['w_E'].append(wE); history['w_S'].append(wS); history['w_W'].append(wW)

        # Agent Decision
        qs = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in all_detectors])
        ws = np.array([traci.lane.getWaitingTime(l) for l in lanes_N+lanes_E+lanes_S+lanes_W])
        sps = np.zeros(8)
        ph_oh = np.zeros(PHASE_DIM); ph_oh[curr_phase] = 1.0
        state = np.concatenate([qs, ws, np.full(4, 0.25), sps, ph_oh])

        state_t = torch.tensor(state, dtype=torch.float32).to(device).unsqueeze(0)
        with torch.no_grad():
            p_logits, mu, _, _ = model(state_t)
            next_phase = torch.argmax(p_logits).item()
            duration = int(torch.clamp(mu, 10.0, 80.0).item())

        # Execute
        if curr_phase != next_phase:
            traci.trafficlight.setRedYellowGreenState(tls_id, phases[curr_phase]["yellow"])
            for _ in range(2): traci.simulationStep(); record(-1); step += 1
            traci.trafficlight.setRedYellowGreenState(tls_id, "r"*12)
            for _ in range(1): traci.simulationStep(); record(-1); step += 1

        traci.trafficlight.setRedYellowGreenState(tls_id, phases[next_phase]["green"])
        for _ in range(duration): traci.simulationStep(); record(next_phase); step += 1

        curr_phase = next_phase

    traci.close()

    # ==========================================================================
    # PLOTTING
    # ==========================================================================
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15), sharex=True, gridspec_kw={'height_ratios': [2, 2, 1]})

    # 1. Queues
    ax1.plot(history['step'], history['q_N'], label='North Q', color='blue')
    ax1.plot(history['step'], history['q_S'], label='South Q', color='green')
    ax1.plot(history['step'], history['q_E'], label='East Q', color='red', linestyle=':')
    ax1.plot(history['step'], history['q_W'], label='West Q', color='orange', linestyle=':')
    ax1.set_ylabel('Queue (Veh)', fontsize=12)
    ax1.set_title('1. Spatial Congestion (Queue Length)', fontsize=14)
    ax1.legend(loc='upper left')
    ax1.grid(True, alpha=0.3)

    # 2. Wait Times (THE NEW METRIC)
    ax2.plot(history['step'], history['w_N'], label='North Wait', color='blue')
    ax2.plot(history['step'], history['w_S'], label='South Wait', color='green')
    ax2.plot(history['step'], history['w_E'], label='East Wait', color='red', linestyle=':')
    ax2.plot(history['step'], history['w_W'], label='West Wait', color='orange', linestyle=':')
    ax2.set_ylabel('Accum. Wait Time (s)', fontsize=12)
    ax2.set_title('2. Driver Pain (Accumulated Wait Time)', fontsize=14)
    ax2.legend(loc='upper left')
    ax2.grid(True, alpha=0.3)

    # 3. Phases
    phases = np.array(history['phase'])
    steps = np.array(history['step'])
    colors = {0: 'blue', 1: 'red', 2: 'cyan', 3: 'magenta', -1: 'gold'}
    for i in range(len(phases)-1):
        ax3.fill_between([steps[i], steps[i+1]], 0, 1, color=colors.get(phases[i], 'gray'), alpha=0.6)

    labels = {0: 'P0: N/S Str', 1: 'P1: E/W Str', 2: 'P2: N/S Left', 3: 'P3: E/W Left', -1: 'Trans'}
    patches = [mpatches.Patch(color=c, label=l) for k, (c, l) in zip(colors.keys(), zip(colors.values(), labels.values()))]
    ax3.legend(handles=patches, loc='center left', bbox_to_anchor=(1, 0.5))
    ax3.set_yticks([])
    ax3.set_xlabel('Time (s)', fontsize=12)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    visualize_complete_picture()
```

```python
#v5 a2c
# @title 15. COMPLETE TRAINER (Squared Reward + Plots + Save/Load)
import os
import sys
import subprocess
import time
import numpy as np
import traci
import traci.constants as tc
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
import matplotlib.pyplot as plt

# ==============================================================================
# 0. SAVING CONFIGURATION
# ==============================================================================
LOAD_MODEL = False          # Set True to resume training
MODEL_DIR = "saved_models"
BEST_MODEL_PATH = os.path.join(MODEL_DIR, "best_model.pth")
LATEST_MODEL_PATH = os.path.join(MODEL_DIR, "latest_model.pth")
os.makedirs(MODEL_DIR, exist_ok=True)

# ==============================================================================
# 1. HYPERPARAMETERS
# ==============================================================================
STATE_DIM = 32
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64
LR = 6e-5
GAMMA = 0.99
ENTROPY_BETA = 0.01

# Action Constraints
MIN_DURATION = 10.0
MAX_DURATION = 60.0

# --- REWARD CONFIG (Squared Wait) ---
REWARD_WEIGHT_DURATION = 0.1
REWARD_SCALING_DIVISOR = 5000.0
REWARD_WEIGHT_QUEUE_DIFF = 0.5
BUMPER_REWARD_BONUS = 10.0
BUMPER_THRESHOLD = 4

# Simulation
SUMO_CFG = "_1.sumocfg"
SIM_DURATION_STEPS = 1500
MAX_EPISODES = 200
BATCH_SIZE = 128

# Traffic System
TLS_ID = "C"
INCOMING_LANES = ["N2C_0", "N2C_1", "E2C_0", "E2C_1", "S2C_0", "S2C_1", "W2C_0", "W2C_1"]
INCOMING_DETECTORS = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]
PHASE_DEFINITIONS = [
    {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
    {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
    {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
    {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
]
ALL_RED_STRING = "r" * 12
global_sumo_process = None

# ==============================================================================
# 2. NEURAL NETWORK
# ==============================================================================
class ActorCriticNet(nn.Module):
    def __init__(self, state_dim, phase_dim):
        super(ActorCriticNet, self).__init__()
        self.backbone = nn.Sequential(
            nn.Linear(state_dim, HIDDEN_1),
            nn.ReLU(),
            nn.Linear(HIDDEN_1, HIDDEN_2),
            nn.ReLU()
        )
        self.actor_phase = nn.Linear(HIDDEN_2, phase_dim)
        self.actor_duration_mu = nn.Linear(HIDDEN_2, 1)
        self.actor_duration_log_std = nn.Linear(HIDDEN_2, 1)
        self.critic_value = nn.Linear(HIDDEN_2, 1)

        with torch.no_grad():
            self.actor_duration_mu.bias.data.fill_(30.0)
            self.actor_duration_log_std.bias.data.fill_(-1.0)

    def forward(self, state):
        x = self.backbone(state)
        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = torch.clamp(self.actor_duration_log_std(x), -20, 2)
        value = self.critic_value(x)
        return phase_logits, mu, log_std, value

# ==============================================================================
# 3. AGENT (With Save/Load)
# ==============================================================================
class A2CAgent:
    def __init__(self):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.net = ActorCriticNet(STATE_DIM, PHASE_DIM).to(self.device)
        self.optimizer = optim.Adam(self.net.parameters(), lr=LR)
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': []}

    def select_action(self, state):
        self.net.eval()
        state_t = torch.tensor(state, dtype=torch.float32).to(self.device).unsqueeze(0)
        with torch.no_grad():
            phase_logits, mu, log_std, value = self.net(state_t)

        phase_dist = distributions.Categorical(logits=phase_logits)
        phase_action = phase_dist.sample()

        std = torch.exp(log_std)
        dur_dist = distributions.Normal(mu, std)
        dur_raw = dur_dist.sample()
        dur_clipped = torch.clamp(dur_raw, MIN_DURATION, MAX_DURATION)

        log_prob = phase_dist.log_prob(phase_action) + dur_dist.log_prob(dur_raw).squeeze()
        return (phase_action.item(), int(dur_clipped.item())), log_prob, value.squeeze()

    def update(self):
        if not self.buffer['states']: return
        states = torch.stack(self.buffer['states']).to(self.device)
        rewards = self.buffer['rewards']
        values = torch.tensor(self.buffer['values']).float().to(self.device)
        log_probs = torch.stack(self.buffer['log_probs']).to(self.device)

        returns = []
        R = values[-1]
        for r in reversed(rewards):
            R = r + GAMMA * R
            returns.insert(0, R)
        returns = torch.tensor(returns).float().to(self.device)

        adv = returns - values
        loss = adv.pow(2).mean() - (log_probs * adv.detach()).mean()

        logits, _, _, _ = self.net(states)
        loss -= ENTROPY_BETA * distributions.Categorical(logits=logits).entropy().mean()

        self.optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(self.net.parameters(), 0.5)
        self.optimizer.step()
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': []}

    # --- SAVE / LOAD FUNCTIONS ---
    def save(self, path):
        torch.save({
            'model': self.net.state_dict(),
            'optim': self.optimizer.state_dict()
        }, path)

    def load(self, path):
        if os.path.exists(path):
            ckpt = torch.load(path, map_location=self.device)
            self.net.load_state_dict(ckpt['model'])
            self.optimizer.load_state_dict(ckpt['optim'])
            print(f" Loaded model from {path}")
            return True
        print(f" Model not found at {path}")
        return False

# ==============================================================================
# 4. ENVIRONMENT
# ==============================================================================
class TrafficAgent:
    def __init__(self, tls_id):
        self.tls_id = tls_id
        self.lanes = INCOMING_LANES
        self.detectors = INCOMING_DETECTORS
        self.phases = PHASE_DEFINITIONS
        self.phase_idx = 0
        self.last_queue_sum = 0
        self.phase_durations = np.zeros(PHASE_DIM)
        self.phase_counts = np.zeros(PHASE_DIM)
        traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[0]["green"])

    def get_state(self):
        queues = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        waits = np.array([traci.lane.getWaitingTime(l) for l in self.lanes])
        speeds = np.array([traci.lanearea.getLastStepMeanSpeed(d) for d in self.detectors])

        norm_q = np.clip(queues / 50.0, 0, 1)
        norm_w = np.clip(waits / 300.0, 0, 1)
        norm_s = np.clip(speeds / 15.0, 0, 1)

        phase_oh = np.zeros(PHASE_DIM); phase_oh[self.phase_idx] = 1.0
        return np.concatenate([norm_q, norm_w, np.full(4, 0.25), norm_s, phase_oh])

    def step(self, action):
        next_phase, duration = action
        self.phase_durations[next_phase] += duration
        self.phase_counts[next_phase] += 1

        if self.phase_idx != next_phase:
            traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[self.phase_idx]["yellow"])
            for _ in range(2): traci.simulationStep()
            traci.trafficlight.setRedYellowGreenState(self.tls_id, ALL_RED_STRING)
            for _ in range(1): traci.simulationStep()

        traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[next_phase]["green"])
        for _ in range(duration): traci.simulationStep()

        self.phase_idx = next_phase
        state = self.get_state()

        # Squared Reward Logic
        current_queues_sum = np.sum([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        all_waits = np.array([traci.lane.getWaitingTime(l) for l in self.lanes])

        squared_wait_penalty = -np.sum(np.square(all_waits))
        r_squared = squared_wait_penalty / REWARD_SCALING_DIVISOR
        r_diff = REWARD_WEIGHT_QUEUE_DIFF * (self.last_queue_sum - current_queues_sum)
        r_bump = BUMPER_REWARD_BONUS if current_queues_sum < BUMPER_THRESHOLD else 0.0

        final_reward = r_squared + r_diff + r_bump
        self.last_queue_sum = current_queues_sum
        return state, final_reward

    def get_metrics(self):
        q = np.mean([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        w = np.sum([traci.lane.getWaitingTime(l) for l in self.lanes])
        s = np.mean([traci.lanearea.getLastStepMeanSpeed(d) for d in self.detectors])
        t = traci.vehicle.getIDCount()
        if np.sum(self.phase_counts) > 0:
            avg_grn = np.divide(self.phase_durations, self.phase_counts, out=np.zeros_like(self.phase_durations, dtype=float), where=self.phase_counts!=0)
        else:
            avg_grn = np.zeros(PHASE_DIM)
        return q, w, s, t, avg_grn.round(1)

# ==============================================================================
# 5. MAIN TRAINING
# ==============================================================================
def train_agent():
    global global_sumo_process
    history = {'rewards': [], 'queues': [], 'waits': [], 'speeds': [], 'throughput': [], 'greens': []}

    # Track best performance (using Queue Length as the primary metric for saving best model)
    best_avg_queue = float('inf')

    # Clean Start
    if global_sumo_process: global_sumo_process.kill()
    sumo_cmd = ["sumo", "-c", SUMO_CFG, "--step-length", "1", "--no-step-log", "--remote-port", "8813"]
    global_sumo_process = subprocess.Popen(sumo_cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(2)
    traci.init(8813)

    agent = A2CAgent()

    # Load if requested
    if LOAD_MODEL:
        agent.load(LATEST_MODEL_PATH)

    try:
        print(f"--- TRAINING START: Squared Reward (Save Dir: {MODEL_DIR}) ---")
        for episode in range(MAX_EPISODES):
            if episode > 0: traci.load(["-c", SUMO_CFG]); traci.simulationStep()

            env = TrafficAgent(TLS_ID)
            state = env.get_state()
            ep_rew = 0
            steps = 0

            m_q, m_w, m_s, m_t = [], [], [], []

            while steps < SIM_DURATION_STEPS:
                action, log_prob, value = agent.select_action(state)
                next_state, reward = env.step(action)

                agent.buffer['states'].append(torch.tensor(state, dtype=torch.float32))
                agent.buffer['rewards'].append(reward)
                agent.buffer['values'].append(value.item())
                agent.buffer['log_probs'].append(log_prob)

                state = next_state
                ep_rew += reward
                steps += (3 + action[1])

                q, w, s, t, _ = env.get_metrics()
                m_q.append(q); m_w.append(w); m_s.append(s); m_t.append(t)

                if len(agent.buffer['states']) >= BATCH_SIZE:
                    agent.update()

            agent.update()

            # --- Stats ---
            avg_q = np.mean(m_q)
            tot_w = np.sum(m_w)
            avg_s = np.mean(m_s)
            avg_t = np.mean(m_t)
            _, _, _, _, avg_greens = env.get_metrics()

            history['rewards'].append(ep_rew)
            history['queues'].append(avg_q)
            history['waits'].append(tot_w)
            history['speeds'].append(avg_s)
            history['throughput'].append(avg_t)
            history['greens'].append(avg_greens)

            # --- SAVE CHECKPOINTS ---
            # 1. Save Latest (Always)
            agent.save(LATEST_MODEL_PATH)

            # 2. Save Best (If Queue is lower than best seen)
            save_msg = ""
            if avg_q < best_avg_queue and episode > 10:
                best_avg_queue = avg_q
                agent.save(BEST_MODEL_PATH)
                save_msg = " [ BEST MODEL SAVED]"

            print(f"Ep {episode+1}: Rew {ep_rew:.0f} | AvgQ {avg_q:.1f} | Wait {tot_w:.0f} | Spd {avg_s:.2f}{save_msg}")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        traci.close()
        if global_sumo_process: global_sumo_process.kill()

    # --- PLOTTING ---
    print("\n--- Saving Plots ---")
    episodes = range(1, len(history['rewards'])+1)

    plt.figure(); plt.plot(episodes, history['rewards'], color='green'); plt.title("Total Reward"); plt.savefig("plot_rewards.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['queues'], color='red'); plt.title("Avg Queue Length"); plt.savefig("plot_queues.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['waits'], color='orange'); plt.title("Total Wait Indicator"); plt.savefig("plot_waits.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['speeds'], color='blue'); plt.title("Avg Speed"); plt.savefig("plot_speeds.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['throughput'], color='purple'); plt.title("Throughput"); plt.savefig("plot_throughput.png"); plt.close()

    plt.figure()
    grn = np.array(history['greens'])
    if len(grn) > 0:
        for i in range(4): plt.plot(episodes, grn[:, i], label=f'Phase {i}')
    plt.legend(); plt.title("Green Time per Phase"); plt.savefig("plot_greens.png"); plt.close()

    print("All plots saved. Training complete.")

if __name__ == "__main__":
    train_agent()
```

```python
# @title 16. TEST SIMULATION (Gen 3 Model)
import os
import sys
import torch
import traci
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ==============================================================================
# 1. CONFIGURATION
# ==============================================================================
# Path to your trained model
MODEL_PATH = "saved_models/best_model.pth"
# Fallback to latest if best doesn't exist
if not os.path.exists(MODEL_PATH):
    MODEL_PATH = "saved_models/latest_model.pth"

SUMO_BINARY = "sumo" # Headless mode
SUMO_CFG = "_1.sumocfg"

# Neural Network Params (Must match training exactly)
STATE_DIM = 32
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64
ACTION_CONTINUOUS_DIM = 1
MIN_DURATION = 10.0
MAX_DURATION = 60.0 # Matching your Gen 3 config

# Lane Config
INCOMING_DETECTORS = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]
LANES_N = ["N2C_0", "N2C_1"]
LANES_E = ["E2C_0", "E2C_1"]
LANES_S = ["S2C_0", "S2C_1"]
LANES_W = ["W2C_0", "W2C_1"]

# ==============================================================================
# 2. MODEL DEFINITION
# ==============================================================================
class ActorCriticNet(torch.nn.Module):
    def __init__(self, state_dim, phase_dim):
        super(ActorCriticNet, self).__init__()
        self.backbone = torch.nn.Sequential(
            torch.nn.Linear(state_dim, HIDDEN_1),
            torch.nn.ReLU(),
            torch.nn.Linear(HIDDEN_1, HIDDEN_2),
            torch.nn.ReLU()
        )
        self.actor_phase = torch.nn.Linear(HIDDEN_2, phase_dim)
        self.actor_duration_mu = torch.nn.Linear(HIDDEN_2, ACTION_CONTINUOUS_DIM)
        self.actor_duration_log_std = torch.nn.Linear(HIDDEN_2, ACTION_CONTINUOUS_DIM)
        self.critic_value = torch.nn.Linear(HIDDEN_2, 1)

    def forward(self, state):
        x = self.backbone(state)
        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = self.actor_duration_log_std(x)
        value = self.critic_value(x)
        return phase_logits, mu, log_std, value

# ==============================================================================
# 3. SIMULATION RUNNER
# ==============================================================================
def run_simulation():
    # --- Setup ---
    device = torch.device("cpu")
    model = ActorCriticNet(STATE_DIM, PHASE_DIM).to(device)

    # --- Load Model ---
    if not os.path.exists(MODEL_PATH):
        print(f" Error: Model file not found at {MODEL_PATH}")
        return

    try:
        checkpoint = torch.load(MODEL_PATH, map_location=device)
        if 'model' in checkpoint:
            model.load_state_dict(checkpoint['model'])
        elif 'model_state_dict' in checkpoint:
            model.load_state_dict(checkpoint['model_state_dict'])
        else:
            model.load_state_dict(checkpoint)
        print(f" Model loaded successfully from {MODEL_PATH}")
    except Exception as e:
        print(f" Error loading model weights: {e}")
        return

    model.eval()

    # --- Start SUMO ---
    try: traci.close()
    except: pass

    print(">>> Starting SUMO (Headless)...")
    traci.start([SUMO_BINARY, "-c", SUMO_CFG, "--step-length", "1", "--no-step-log"])

    # --- Data Containers ---
    history = {
        'step': [], 'phase': [],
        'q_N': [], 'q_E': [], 'q_S': [], 'q_W': [],
        'w_N': [], 'w_E': [], 'w_S': [], 'w_W': []
    }

    # --- Simulation Loop ---
    step = 0
    tls_id = "C"
    phases = [
        {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
        {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
        {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
        {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
    ]

    curr_phase = 0
    traci.trafficlight.setRedYellowGreenState(tls_id, phases[0]["green"])

    print(">>> Running Simulation Loop...")

    while traci.simulation.getMinExpectedNumber() > 0 and step < 4000:

        # 1. Helper to record detailed metrics per direction
        def record_metrics(ph):
            # Queues
            qN = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_N])
            qE = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_E])
            qS = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_S])
            qW = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_W])

            # Wait Times
            wN = np.sum([traci.lane.getWaitingTime(l) for l in LANES_N])
            wE = np.sum([traci.lane.getWaitingTime(l) for l in LANES_E])
            wS = np.sum([traci.lane.getWaitingTime(l) for l in LANES_S])
            wW = np.sum([traci.lane.getWaitingTime(l) for l in LANES_W])

            history['step'].append(step)
            history['phase'].append(ph)
            history['q_N'].append(qN); history['q_E'].append(qE); history['q_S'].append(qS); history['q_W'].append(qW)
            history['w_N'].append(wN); history['w_E'].append(wE); history['w_S'].append(wS); history['w_W'].append(wW)

        # 2. Get State
        qs = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in INCOMING_DETECTORS])
        all_lanes = LANES_N + LANES_E + LANES_S + LANES_W
        ws = np.array([traci.lane.getWaitingTime(l) for l in all_lanes])
        sps = np.array([traci.lanearea.getLastStepMeanSpeed(d) for d in INCOMING_DETECTORS])

        # Apply Normalization (Must match training!)
        norm_q = np.clip(qs / 50.0, 0, 1)
        norm_w = np.clip(ws / 300.0, 0, 1)
        norm_s = np.clip(sps / 15.0, 0, 1)

        ph_oh = np.zeros(PHASE_DIM); ph_oh[curr_phase] = 1.0

        state = np.concatenate([norm_q, norm_w, np.full(4, 0.25), norm_s, ph_oh])

        # 3. Model Inference
        state_t = torch.tensor(state, dtype=torch.float32).to(device).unsqueeze(0)
        with torch.no_grad():
            p_logits, mu, _, _ = model(state_t)
            # Use argmax for deterministic testing
            next_phase = torch.argmax(p_logits).item()
            duration = int(torch.clamp(mu, MIN_DURATION, MAX_DURATION).item())

        # 4. Execute Phase Change in SUMO
        if curr_phase != next_phase:
            # Yellow
            traci.trafficlight.setRedYellowGreenState(tls_id, phases[curr_phase]["yellow"])
            for _ in range(2):
                traci.simulationStep(); record_metrics(-1); step += 1
            # Red
            traci.trafficlight.setRedYellowGreenState(tls_id, "r"*len(phases[0]["green"]))
            for _ in range(1):
                traci.simulationStep(); record_metrics(-1); step += 1

        # Green
        traci.trafficlight.setRedYellowGreenState(tls_id, phases[next_phase]["green"])
        for _ in range(duration):
            traci.simulationStep(); record_metrics(next_phase); step += 1

        curr_phase = next_phase

    traci.close()
    print(">>> Simulation Complete.")

    # ==========================================================================
    # 4. PLOTTING RESULTS
    # ==========================================================================
    print(">>> Generating Diagnostic Plots...")
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15), sharex=True, gridspec_kw={'height_ratios': [2, 2, 1]})

    # Graph 1: Queue Lengths
    ax1.plot(history['step'], history['q_N'], label='North Q', color='blue', alpha=0.8)
    ax1.plot(history['step'], history['q_S'], label='South Q', color='green', alpha=0.8)
    ax1.plot(history['step'], history['q_E'], label='East Q', color='red', linestyle=':', alpha=0.8)
    ax1.plot(history['step'], history['q_W'], label='West Q', color='orange', linestyle=':', alpha=0.8)
    ax1.set_ylabel('Queue Length (Veh)', fontsize=12)
    ax1.set_title(f'1. Congestion Analysis (Queues) - Gen 3 Model', fontsize=14)
    ax1.legend(loc='upper left')
    ax1.grid(True, alpha=0.3)

    # Graph 2: Accumulated Wait Times
    ax2.plot(history['step'], history['w_N'], label='North Wait', color='blue')
    ax2.plot(history['step'], history['w_S'], label='South Wait', color='green')
    ax2.plot(history['step'], history['w_E'], label='East Wait', color='red', linestyle=':')
    ax2.plot(history['step'], history['w_W'], label='West Wait', color='orange', linestyle=':')
    ax2.set_ylabel('Accum. Wait Time (s)', fontsize=12)
    ax2.set_title('2. Fairness Analysis (Wait Times)', fontsize=14)
    ax2.legend(loc='upper left')
    ax2.grid(True, alpha=0.3)

    # Graph 3: Phase Allocation
    phases = np.array(history['phase'])
    steps = np.array(history['step'])
    colors = {0: 'blue', 1: 'red', 2: 'cyan', 3: 'magenta', -1: 'gold'}

    for i in range(len(phases)-1):
        ax3.fill_between([steps[i], steps[i+1]], 0, 1, color=colors.get(phases[i], 'gray'), alpha=0.6)

    labels = {0: 'P0: N/S Str', 1: 'P1: E/W Str', 2: 'P2: N/S Left', 3: 'P3: E/W Left', -1: 'Transition'}
    patches = [mpatches.Patch(color=c, label=l) for k, (c, l) in zip(colors.keys(), zip(colors.values(), labels.values()))]
    ax3.legend(handles=patches, loc='center left', bbox_to_anchor=(1, 0.5))
    ax3.set_yticks([])
    ax3.set_xlabel('Simulation Time (s)', fontsize=12)
    ax3.set_title('3. Agent Strategy (Phase Sequence)', fontsize=14)

    plt.tight_layout()
    plt.savefig("test_simulation_results.png")
    plt.show()

if __name__ == "__main__":
    run_simulation()
```

```python
# v6 a2c
# @title 17. GEN 4 TRAINING: PROACTIVE AGENT (Time Memory + Squared Reward)
import os
import sys
import subprocess
import time
import numpy as np
import traci
import traci.constants as tc
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
import matplotlib.pyplot as plt

# ==============================================================================
# 0. SAVING CONFIGURATION
# ==============================================================================
LOAD_MODEL = False          # Set True to resume
MODEL_DIR = "saved_models_gen4"
BEST_MODEL_PATH = os.path.join(MODEL_DIR, "best_model_gen4.pth")
LATEST_MODEL_PATH = os.path.join(MODEL_DIR, "latest_model_gen4.pth")
os.makedirs(MODEL_DIR, exist_ok=True)

# ==============================================================================
# 1. HYPERPARAMETERS (GEN 4)
# ==============================================================================
# --- State Dimension increased for "Time Since Green" ---
STATE_DIM = 36  # 32 original + 4 phase timers
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64
LR = 5e-5
GAMMA = 0.995
ENTROPY_BETA = 0.02

# --- Action Constraints ---
MIN_DURATION = 10.0
MAX_DURATION = 60.0

# --- REWARD CONFIG (Squared Fairness) ---
REWARD_WEIGHT_DURATION = 0.0
REWARD_SCALING_DIVISOR = 6000.0
REWARD_WEIGHT_QUEUE_DIFF = 0.5
BUMPER_REWARD_BONUS = 10.0
BUMPER_THRESHOLD = 4

# --- Simulation ---
SUMO_CFG = "_1.sumocfg"
SIM_DURATION_STEPS = 1800
MAX_EPISODES = 600
BATCH_SIZE = 128

# --- Traffic System ---
TLS_ID = "C"
INCOMING_LANES = ["N2C_0", "N2C_1", "E2C_0", "E2C_1", "S2C_0", "S2C_1", "W2C_0", "W2C_1"]
INCOMING_DETECTORS = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]
PHASE_DEFINITIONS = [
    {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
    {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
    {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
    {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
]
ALL_RED_STRING = "r" * 12
global_sumo_process = None

# ==============================================================================
# 2. NEURAL NETWORK
# ==============================================================================
class ActorCriticNet(nn.Module):
    def __init__(self, state_dim, phase_dim):
        super(ActorCriticNet, self).__init__()
        self.backbone = nn.Sequential(
            nn.Linear(state_dim, HIDDEN_1),
            nn.ReLU(),
            nn.Linear(HIDDEN_1, HIDDEN_2),
            nn.ReLU()
        )
        self.actor_phase = nn.Linear(HIDDEN_2, phase_dim)
        self.actor_duration_mu = nn.Linear(HIDDEN_2, 1)
        self.actor_duration_log_std = nn.Linear(HIDDEN_2, 1)
        self.critic_value = nn.Linear(HIDDEN_2, 1)

        with torch.no_grad():
            self.actor_duration_mu.bias.data.fill_(30.0)
            self.actor_duration_log_std.bias.data.fill_(-1.0)

    def forward(self, state):
        x = self.backbone(state)
        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = torch.clamp(self.actor_duration_log_std(x), -20, 2)
        value = self.critic_value(x)
        return phase_logits, mu, log_std, value

# ==============================================================================
# 3. AGENT
# ==============================================================================
class A2CAgent:
    def __init__(self):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.net = ActorCriticNet(STATE_DIM, PHASE_DIM).to(self.device)
        self.optimizer = optim.Adam(self.net.parameters(), lr=LR)
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': []}

    def select_action(self, state):
        self.net.eval()
        state_t = torch.tensor(state, dtype=torch.float32).to(self.device).unsqueeze(0)
        with torch.no_grad():
            phase_logits, mu, log_std, value = self.net(state_t)

        phase_dist = distributions.Categorical(logits=phase_logits)
        phase_action = phase_dist.sample()

        std = torch.exp(log_std)
        dur_dist = distributions.Normal(mu, std)
        dur_raw = dur_dist.sample()
        dur_clipped = torch.clamp(dur_raw, MIN_DURATION, MAX_DURATION)

        log_prob = phase_dist.log_prob(phase_action) + dur_dist.log_prob(dur_raw).squeeze()
        return (phase_action.item(), int(dur_clipped.item())), log_prob, value.squeeze()

    def update(self):
        if not self.buffer['states']: return
        states = torch.stack(self.buffer['states']).to(self.device)
        rewards = self.buffer['rewards']
        values = torch.tensor(self.buffer['values']).float().to(self.device)
        log_probs = torch.stack(self.buffer['log_probs']).to(self.device)

        returns = []
        R = values[-1]
        for r in reversed(rewards):
            R = r + GAMMA * R
            returns.insert(0, R)
        returns = torch.tensor(returns).float().to(self.device)

        adv = returns - values
        loss = adv.pow(2).mean() - (log_probs * adv.detach()).mean()

        logits, _, _, _ = self.net(states)
        loss -= ENTROPY_BETA * distributions.Categorical(logits=logits).entropy().mean()

        self.optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(self.net.parameters(), 0.5)
        self.optimizer.step()
        self.buffer = {'states': [], 'rewards': [], 'values': [], 'log_probs': []}

    def save(self, path):
        torch.save({'model': self.net.state_dict(), 'optim': self.optimizer.state_dict()}, path)

    def load(self, path):
        if os.path.exists(path):
            ckpt = torch.load(path, map_location=self.device)
            self.net.load_state_dict(ckpt['model'])
            self.optimizer.load_state_dict(ckpt['optim'])
            print(f" Loaded model from {path}")
            return True
        print(f" Model not found at {path}")
        return False

# ==============================================================================
# 4. ENVIRONMENT (Proactive Logic)
# ==============================================================================
class TrafficAgent:
    def __init__(self, tls_id):
        self.tls_id = tls_id
        self.lanes = INCOMING_LANES
        self.detectors = INCOMING_DETECTORS
        self.phases = PHASE_DEFINITIONS
        self.phase_idx = 0
        self.last_queue_sum = 0
        self.yellow_time = 2
        self.all_red_time = 1
        self.phase_durations = np.zeros(PHASE_DIM)
        self.phase_counts = np.zeros(PHASE_DIM)

        # --- GEN 4: MEMORY ---
        # Tracks seconds since each phase was last green
        self.time_since_last_green = np.zeros(PHASE_DIM, dtype=float)

        traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[0]["green"])

    def get_state(self):
        queues = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        waits = np.array([traci.lane.getWaitingTime(l) for l in self.lanes])
        speeds = np.array([traci.lanearea.getLastStepMeanSpeed(d) for d in self.detectors])

        norm_q = np.clip(queues / 50.0, 0, 1)
        norm_w = np.clip(waits / 300.0, 0, 1)
        norm_s = np.clip(speeds / 15.0, 0, 1)
        phase_oh = np.zeros(PHASE_DIM); phase_oh[self.phase_idx] = 1.0

        # --- GEN 4: ADD TIMER TO STATE ---
        # Normalize by 100 seconds (assuming >100s is critical danger)
        norm_timer = np.clip(self.time_since_last_green / 100.0, 0, 1)

        return np.concatenate([norm_q, norm_w, np.full(4, 0.25), norm_s, phase_oh, norm_timer])

    def step(self, action):
        next_phase, duration = action

        # --- GEN 4: UPDATE TIMERS ---
        # Increment timer for ALL phases by the total duration of this step
        total_step_time = duration
        if self.phase_idx != next_phase:
            total_step_time += (self.yellow_time + self.all_red_time)

        self.time_since_last_green += total_step_time
        # Reset timer for the phase that is about to turn GREEN
        self.time_since_last_green[next_phase] = 0

        # Metrics
        self.phase_durations[next_phase] += duration
        self.phase_counts[next_phase] += 1

        # Execution
        if self.phase_idx != next_phase:
            traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[self.phase_idx]["yellow"])
            for _ in range(self.yellow_time): traci.simulationStep()
            traci.trafficlight.setRedYellowGreenState(self.tls_id, ALL_RED_STRING)
            for _ in range(self.all_red_time): traci.simulationStep()

        traci.trafficlight.setRedYellowGreenState(self.tls_id, self.phases[next_phase]["green"])
        for _ in range(duration): traci.simulationStep()

        self.phase_idx = next_phase
        state = self.get_state()

        # Squared Reward
        current_queues_sum = np.sum([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        all_waits = np.array([traci.lane.getWaitingTime(l) for l in self.lanes])

        squared_wait_penalty = -np.sum(np.square(all_waits))
        r_squared = squared_wait_penalty / REWARD_SCALING_DIVISOR
        r_diff = REWARD_WEIGHT_QUEUE_DIFF * (self.last_queue_sum - current_queues_sum)
        r_bump = BUMPER_REWARD_BONUS if current_queues_sum < BUMPER_THRESHOLD else 0.0

        final_reward = r_squared + r_diff + r_bump
        self.last_queue_sum = current_queues_sum
        return state, final_reward

    def get_metrics(self):
        q = np.mean([traci.lanearea.getLastStepHaltingNumber(d) for d in self.detectors])
        w = np.sum([traci.lane.getWaitingTime(l) for l in self.lanes])
        s = np.mean([traci.lanearea.getLastStepMeanSpeed(d) for d in self.detectors])
        t = traci.vehicle.getIDCount()
        if np.sum(self.phase_counts) > 0:
            avg_grn = np.divide(self.phase_durations, self.phase_counts, out=np.zeros_like(self.phase_durations, dtype=float), where=self.phase_counts!=0)
        else:
            avg_grn = np.zeros(PHASE_DIM)
        return q, w, s, t, avg_grn.round(1)

# ==============================================================================
# 5. MAIN TRAINING LOOP
# ==============================================================================
def train_agent():
    global global_sumo_process
    history = {'rewards': [], 'queues': [], 'waits': [], 'speeds': [], 'throughput': [], 'greens': []}
    best_avg_queue = float('inf')

    if global_sumo_process: global_sumo_process.kill()
    sumo_cmd = ["sumo", "-c", SUMO_CFG, "--step-length", "1", "--no-step-log", "--remote-port", "8813"]
    global_sumo_process = subprocess.Popen(sumo_cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(2)
    traci.init(8813)

    agent = A2CAgent()
    if LOAD_MODEL: agent.load(LATEST_MODEL_PATH)

    try:
        print(f"--- TRAINING START: GEN 4 PROACTIVE (Save Dir: {MODEL_DIR}) ---")
        for episode in range(MAX_EPISODES):
            if episode > 0: traci.load(["-c", SUMO_CFG]); traci.simulationStep()

            env = TrafficAgent(TLS_ID)
            state = env.get_state()
            ep_rew = 0
            steps = 0

            m_q, m_w, m_s, m_t = [], [], [], []

            while steps < SIM_DURATION_STEPS:
                action, log_prob, value = agent.select_action(state)
                next_state, reward = env.step(action)

                agent.buffer['states'].append(torch.tensor(state, dtype=torch.float32))
                agent.buffer['rewards'].append(reward)
                agent.buffer['values'].append(value.item())
                agent.buffer['log_probs'].append(log_prob)

                state = next_state
                ep_rew += reward
                steps += (3 + action[1])

                q, w, s, t, _ = env.get_metrics()
                m_q.append(q); m_w.append(w); m_s.append(s); m_t.append(t)

                if len(agent.buffer['states']) >= BATCH_SIZE:
                    agent.update()

            agent.update()

            # Stats
            avg_q = np.mean(m_q)
            tot_w = np.sum(m_w)
            avg_s = np.mean(m_s)
            avg_t = np.mean(m_t)
            _, _, _, _, avg_greens = env.get_metrics()

            history['rewards'].append(ep_rew)
            history['queues'].append(avg_q)
            history['waits'].append(tot_w)
            history['speeds'].append(avg_s)
            history['throughput'].append(avg_t)
            history['greens'].append(avg_greens)

            # Save
            agent.save(LATEST_MODEL_PATH)
            save_msg = ""
            if avg_q < best_avg_queue and episode > 10:
                best_avg_queue = avg_q
                agent.save(BEST_MODEL_PATH)
                save_msg = " [ BEST SAVED]"

            # Full Analysis Print
            print(f"Ep {episode+1}: Rew {ep_rew:.0f} | AvgQ {avg_q:.1f} | Wait {tot_w:.0f} | Spd {avg_s:.2f} | Thru {avg_t:.1f} | Grn {avg_greens}{save_msg}")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        traci.close()
        if global_sumo_process: global_sumo_process.kill()

    # Saving Plots
    print("\n--- Generating Analysis Plots ---")
    episodes = range(1, len(history['rewards'])+1)

    plt.figure(); plt.plot(episodes, history['rewards'], color='green'); plt.title("Total Reward"); plt.savefig("plot_rewards.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['queues'], color='red'); plt.title("Avg Queue Length"); plt.savefig("plot_queues.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['waits'], color='orange'); plt.title("Total Wait Indicator"); plt.savefig("plot_waits.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['speeds'], color='blue'); plt.title("Avg Speed"); plt.savefig("plot_speeds.png"); plt.close()
    plt.figure(); plt.plot(episodes, history['throughput'], color='purple'); plt.title("Throughput"); plt.savefig("plot_throughput.png"); plt.close()

    plt.figure()
    grn = np.array(history['greens'])
    if len(grn) > 0:
        for i in range(4): plt.plot(episodes, grn[:, i], label=f'Phase {i}')
    plt.legend(); plt.title("Avg Green Time per Phase"); plt.savefig("plot_greens.png"); plt.close()

    print("Done. Plots saved to current directory.")

if __name__ == "__main__":
    train_agent()
```

```python
# @title 18. MASTER TEST SCRIPT: FULL DIAGNOSTIC DASHBOARD
import os
import sys
import torch
import traci
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ==============================================================================
# 1. CONFIGURATION
# ==============================================================================
MODEL_PATH = "saved_models_gen4/best_model_gen4.pth"
# Fallback
if not os.path.exists(MODEL_PATH):
    MODEL_PATH = "saved_models_gen4/latest_model_gen4.pth"

SUMO_BINARY = "sumo"
SUMO_CFG = "_1.sumocfg"

# Gen 4 Params (Must match training)
STATE_DIM = 36  # 32 + 4 timers
PHASE_DIM = 4
HIDDEN_1 = 128
HIDDEN_2 = 64
MIN_DURATION = 10.0
MAX_DURATION = 60.0

# Detectors
INCOMING_DETECTORS = ["flow-N2C-0", "flow-N2C-1", "flow-E2C-0", "flow-E2C-1", "flow-S2C-0", "flow-S2C-1", "flow-W2C-0", "flow-W2C-1"]
LANES_N = ["N2C_0", "N2C_1"]
LANES_E = ["E2C_0", "E2C_1"]
LANES_S = ["S2C_0", "S2C_1"]
LANES_W = ["W2C_0", "W2C_1"]

# ==============================================================================
# 2. MODEL ARCHITECTURE
# ==============================================================================
class ActorCriticNet(torch.nn.Module):
    def __init__(self, state_dim, phase_dim):
        super(ActorCriticNet, self).__init__()
        self.backbone = torch.nn.Sequential(
            torch.nn.Linear(state_dim, HIDDEN_1),
            torch.nn.ReLU(),
            torch.nn.Linear(HIDDEN_1, HIDDEN_2),
            torch.nn.ReLU()
        )
        self.actor_phase = torch.nn.Linear(HIDDEN_2, phase_dim)
        self.actor_duration_mu = torch.nn.Linear(HIDDEN_2, 1)
        self.actor_duration_log_std = torch.nn.Linear(HIDDEN_2, 1)
        self.critic_value = torch.nn.Linear(HIDDEN_2, 1)

    def forward(self, state):
        x = self.backbone(state)
        phase_logits = self.actor_phase(x)
        mu = self.actor_duration_mu(x)
        log_std = self.actor_duration_log_std(x)
        return phase_logits, mu, log_std, self.critic_value(x)

# ==============================================================================
# 3. SIMULATION RUNNER
# ==============================================================================
def run_comprehensive_test():
    # --- Init Model ---
    device = torch.device("cpu")
    model = ActorCriticNet(STATE_DIM, PHASE_DIM).to(device)

    if not os.path.exists(MODEL_PATH):
        print(f" Error: Model not found at {MODEL_PATH}")
        return

    try:
        checkpoint = torch.load(MODEL_PATH, map_location=device)
        if 'model' in checkpoint: model.load_state_dict(checkpoint['model'])
        elif 'model_state_dict' in checkpoint: model.load_state_dict(checkpoint['model_state_dict'])
        else: model.load_state_dict(checkpoint)
        print(f" Loaded Model: {MODEL_PATH}")
    except Exception as e:
        print(f" Error loading weights: {e}")
        return

    model.eval()

    # --- Start SUMO ---
    try: traci.close()
    except: pass
    traci.start([SUMO_BINARY, "-c", SUMO_CFG, "--step-length", "1", "--no-step-log"])

    # --- State Tracking ---
    time_since_last_green = np.zeros(PHASE_DIM)
    curr_phase = 0

    phases = [
        {"green": "gGrrrrgGrrrr", "yellow": "yyrrrryyrrrr"},
        {"green": "rrrgGrrrrgGr", "yellow": "rrryyrrrryyr"},
        {"green": "rrrrrGrrrrrG", "yellow": "rrrrryrrrrry"},
        {"green": "rrGrrrrrGrrr", "yellow": "rrYrrrrrrYrr"}
    ]
    traci.trafficlight.setRedYellowGreenState("C", phases[0]["green"])

    # --- History Container ---
    history = {
        'step': [], 'phase': [],
        'q_N': [], 'q_E': [], 'q_S': [], 'q_W': [], 'q_Tot': [],
        'w_N': [], 'w_E': [], 'w_S': [], 'w_W': [],
        'speed': [], 'throughput': []
    }

    print(">>> Running Simulation...")
    step = 0
    while traci.simulation.getMinExpectedNumber() > 0 and step < 3600:

        # --- 1. Record Data Helper ---
        def record(ph):
            # Queues
            qN = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_N])
            qE = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_E])
            qS = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_S])
            qW = np.sum([traci.lanearea.getLastStepHaltingNumber(f"flow-{l.replace('_','-')}") for l in LANES_W])

            # Wait
            wN = np.sum([traci.lane.getWaitingTime(l) for l in LANES_N])
            wE = np.sum([traci.lane.getWaitingTime(l) for l in LANES_E])
            wS = np.sum([traci.lane.getWaitingTime(l) for l in LANES_S])
            wW = np.sum([traci.lane.getWaitingTime(l) for l in LANES_W])

            # Speed & Throughput
            spd = np.mean([traci.lanearea.getLastStepMeanSpeed(d) for d in INCOMING_DETECTORS])
            thru = traci.simulation.getDepartedNumber()

            history['step'].append(step)
            history['phase'].append(ph)
            history['q_N'].append(qN); history['q_E'].append(qE); history['q_S'].append(qS); history['q_W'].append(qW)
            history['q_Tot'].append(qN+qE+qS+qW)
            history['w_N'].append(wN); history['w_E'].append(wE); history['w_S'].append(wS); history['w_W'].append(wW)
            history['speed'].append(spd)
            # Cumulative throughput logic
            prev_thru = history['throughput'][-1] if len(history['throughput']) > 0 else 0
            history['throughput'].append(prev_thru + thru)

        # --- 2. Construct State (GEN 4 LOGIC) ---
        qs = np.array([traci.lanearea.getLastStepHaltingNumber(d) for d in INCOMING_DETECTORS])
        all_lanes = LANES_N + LANES_E + LANES_S + LANES_W
        ws = np.array([traci.lane.getWaitingTime(l) for l in all_lanes])
        sps = np.array([traci.lanearea.getLastStepMeanSpeed(d) for d in INCOMING_DETECTORS])

        norm_q = np.clip(qs / 50.0, 0, 1)
        norm_w = np.clip(ws / 300.0, 0, 1)
        norm_s = np.clip(sps / 15.0, 0, 1)
        ph_oh = np.zeros(PHASE_DIM); ph_oh[curr_phase] = 1.0
        norm_timer = np.clip(time_since_last_green / 100.0, 0, 1) # The Missing Piece

        state = np.concatenate([norm_q, norm_w, np.full(4, 0.25), norm_s, ph_oh, norm_timer])

        # --- 3. Inference ---
        state_t = torch.tensor(state, dtype=torch.float32).to(device).unsqueeze(0)
        with torch.no_grad():
            p_logits, mu, _, _ = model(state_t)
            next_phase = torch.argmax(p_logits).item()
            duration = int(torch.clamp(mu, MIN_DURATION, MAX_DURATION).item())

        # --- 4. Update Gen 4 Timers ---
        # Increment all timers
        total_action_time = duration
        if curr_phase != next_phase: total_action_time += 3 # Yellow + Red
        time_since_last_green += total_action_time
        # Reset the one we are choosing
        time_since_last_green[next_phase] = 0

        # --- 5. Execute ---
        if curr_phase != next_phase:
            traci.trafficlight.setRedYellowGreenState("C", phases[curr_phase]["yellow"])
            for _ in range(2): traci.simulationStep(); record(-1); step += 1
            traci.trafficlight.setRedYellowGreenState("C", "r"*12)
            for _ in range(1): traci.simulationStep(); record(-1); step += 1

        traci.trafficlight.setRedYellowGreenState("C", phases[next_phase]["green"])
        for _ in range(duration):
            traci.simulationStep(); record(next_phase); step += 1

        curr_phase = next_phase

    traci.close()
    print(">>> Simulation Complete.")

    # ==========================================================================
    # 6. MASTER DASHBOARD PLOT
    # ==========================================================================
    print(">>> Generating Dashboard...")
    fig, axs = plt.subplots(5, 1, figsize=(14, 20), sharex=True, gridspec_kw={'height_ratios': [2, 2, 1, 1, 1]})
    fig.suptitle(f'Master Diagnostic Dashboard: Gen 4 Model', fontsize=18, weight='bold')

    # 1. Queues (Broken Down)
    axs[0].plot(history['step'], history['q_Tot'], color='black', linewidth=2, alpha=0.3, label='Total')
    axs[0].plot(history['step'], history['q_N'], label='North', color='blue')
    axs[0].plot(history['step'], history['q_S'], label='South', color='green')
    axs[0].plot(history['step'], history['q_E'], label='East', color='red', linestyle=':')
    axs[0].plot(history['step'], history['q_W'], label='West', color='orange', linestyle=':')
    axs[0].set_ylabel('Queue (Veh)', fontsize=12, weight='bold')
    axs[0].set_title('1. Queue Balance (Congestion)', fontsize=14)
    axs[0].legend(loc='upper left', ncol=5)
    axs[0].grid(True, alpha=0.3)

    # 2. Wait Times
    axs[1].plot(history['step'], history['w_N'], label='North', color='blue')
    axs[1].plot(history['step'], history['w_S'], label='South', color='green')
    axs[1].plot(history['step'], history['w_E'], label='East', color='red', linestyle=':')
    axs[1].plot(history['step'], history['w_W'], label='West', color='orange', linestyle=':')
    axs[1].set_ylabel('Accum. Wait (s)', fontsize=12, weight='bold')
    axs[1].set_title('2. Driver Pain (Fairness Analysis)', fontsize=14)
    axs[1].grid(True, alpha=0.3)

    # 3. Average Speed
    axs[2].plot(history['step'], history['speed'], color='teal', alpha=0.8)
    axs[2].set_ylabel('Speed (m/s)', fontsize=12, weight='bold')
    axs[2].set_title('3. Traffic Flow Velocity', fontsize=14)
    axs[2].grid(True, alpha=0.3)

    # 4. Throughput
    axs[3].plot(history['step'], history['throughput'], color='purple', linewidth=2)
    axs[3].set_ylabel('Vehicles', fontsize=12, weight='bold')
    axs[3].set_title('4. Cumulative System Throughput', fontsize=14)
    axs[3].grid(True, alpha=0.3)

    # 5. Phase Gantt Chart
    phases = np.array(history['phase'])
    steps = np.array(history['step'])
    colors = {0: 'blue', 1: 'red', 2: 'cyan', 3: 'magenta', -1: 'gold'}
    labels = {0: 'N/S Str', 1: 'E/W Str', 2: 'N/S Left', 3: 'E/W Left', -1: 'Trans'}

    for i in range(len(phases)-1):
        axs[4].fill_between([steps[i], steps[i+1]], 0, 1, color=colors.get(phases[i], 'gray'), alpha=0.8)

    patches = [mpatches.Patch(color=c, label=l) for k, (c, l) in zip(colors.keys(), zip(colors.values(), labels.values()))]
    axs[4].legend(handles=patches, loc='center left', bbox_to_anchor=(1, 0.5))
    axs[4].set_yticks([])
    axs[4].set_xlabel('Simulation Time (Seconds)', fontsize=14, weight='bold')
    axs[4].set_title('5. Agent Strategy Timeline', fontsize=14)

    plt.tight_layout(rect=[0, 0.03, 1, 0.97])
    plt.savefig("master_dashboard.png")
    plt.show()

if __name__ == "__main__":
    run_comprehensive_test()
```
